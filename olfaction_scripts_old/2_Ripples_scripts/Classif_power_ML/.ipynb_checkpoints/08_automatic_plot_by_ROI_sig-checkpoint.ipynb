```python
"""
PLOT significant data automatically from .csv files
"""
from brainpipe.system import study
import matplotlib.pyplot as plt
from itertools import product
from os import listdir
from os.path import join
import pandas as pd
import numpy as np
from brainpipe.visual import BorderPlot, addLines, rmaxis
from matplotlib.ticker import MaxNLocator, ScalarFormatter

st = study('Ripples')
path_rep = join(st.path, 'classified/{}/')
path_res = join(path_rep, '{}_results_th={}_minwin={}/')
mat_file = join(path_rep, 'All_subjects_sources_{}_{}_{}.npz')
plt_save = 'Pow_DA_roi={}_sign={}_su={}_ch={}_th={}_minwin={}.png'

exp, freqs = 'Retrieval_new_rec', ['HFA','low_freq','ripple']
#'Retrieval_new_odors/', 'Retrieval_new_rec/', 'Encoding/'
th, min_wins, conds = '0.05', ['1'], ['low','high']

for freq, win in product(freqs,min_wins):
    rois = listdir(path_res.format(exp,freq,th,win))
    for roi in rois:
        csv_list = [f for f in listdir(path_res.format(exp,freq,th,win)+roi+'/')\
                            if f.endswith('.csv')]
        for csv_file in csv_list:
            sign = csv_file.split('sign=')[1][:3]
            df_res = pd.read_csv(path_res.format(exp,freq,th,win)+roi+'/'+csv_file)
            mat = np.load(mat_file.format(exp,freq, conds[0], conds[1]),
                         allow_pickle=True)

            for su,ch in zip(df_res['subjects'], df_res['channels']):
                print(roi,su,ch,'th=',th,'win=',win)
                idx_ = [i for i,chan in enumerate(mat['s_channels']) \
                        if (chan == ch) & (mat['su_codes'][i] == su)]

                data_pow = np.concatenate((mat['s_elec_pow0'][idx_],
                                           mat['s_elec_pow1'][idx_]), axis=0)
                print(idx_, data_pow.shape)
                auc = mat['s_da'][idx_]
                th_su = df_res['th_su'].loc[((df_res.subjects == su) & (df_res.channels == ch))]
                th_su = float(th_su)

                xfmt = ScalarFormatter(useMathText=True)
                xfmt.set_powerlimits((0,3))
                fig = plt.figure(1,figsize=(7,7))
                title = freq+' Power & DA for '+roi+' ('+su+' in '+str(ch)+')'
                fig.suptitle(title, fontsize=12)

                plt.subplot(211)
                time = mat['s_time']
                BorderPlot(time, data_pow, y=[0,1], color=['blue','red'],linewidth=2, 
                              ncol=1, xlabel='Time (s)',ylabel = r'Power', legend=conds)
                rmaxis(plt.gca(), ['right', 'top'])
                addLines(plt.gca(), vLines=[0], vColor=['darkgray'], vWidth=[2])
                plt.legend(loc=0, handletextpad=0.1, frameon=False)
                plt.gca().yaxis.set_major_locator(MaxNLocator(3,integer=True))

                #Plot significant time period on power data
                points_sig = [t for idx_t, t in enumerate(time) if auc[0,idx_t]>th_su]
                if len(points_sig) > 1:
                    h = plt.axvspan(points_sig[0], points_sig[-1],
                            color='r', alpha=0.3)

                plt.subplot(212)
                BorderPlot(time, auc, color='black',xlabel='Time (s)', ylim=[0.2,1.], 
                               ylabel='Decoding accuracy (%)',linewidth=2, alpha=0.3)
                rmaxis(plt.gca(), ['right', 'top'])
                addLines(plt.gca(), vLines=[0], vColor=['darkgray'], vWidth=[2])
                plt.gca().yaxis.set_major_locator(MaxNLocator(3,integer=True))
                plt.plot(time, [th_su]*np.ones(len(time)), '--', color='r', linewidth=2)
                plt.savefig(path_res.format(exp,freq,th,win)+roi+'/'+plt_save.format(roi,sign,su,ch,th,win))
                plt.clf()
                plt.close()
```
