### Compute Power across time
    No baseline correction 

```python
from brainpipe.feat.utils._feat import _manageWindow
from brainpipe.feature import power

st = study('Olfacto')

################################################################################################
# Define power settings :
bsl, width, step = None, 358, 51 #358, 51 #[(768,1024),(1024,1280)],[(1280,1536)]
baseline = 'No baseline' #['-500_0'] #['-1500_-1000','-1000_-500'] #'700ms'
################################################################################################
reps = ['Encoding_By_Odor']

for rep in reps:
    path2save = path.join(st.path, 'feature/TPSim_power_data/Power_all_elecs_E_R_by_odor_all_elecsFT/')
    code = 'E' if rep.startswith('Encoding') else 'R'
    ###############################################################################
    if not path.exists(path2save):
        makedirs(path2save)
    ###############################################################################  
    files = [k for k in st.search('bipo_all_noWM_physFT.npz', folder=('database/'+rep+'/'))]
    for fi in files:
        # Dict with all power features
        kwargs = {} # Define an empty dictionnary to save all power parameters
        kwargs['f'] = [[1,3],[3,7],[7,13],[13,30],[30,60],[60,120]] # Frequency vector
        fname = ['delta', 'theta', 'alpha', 'beta', 'gamma_low', 'gamma_high'] # Name of each frequencykwargs['f'] = [[3,7], [8,13], [13,30], [30,120]] # Frequency vector[0.1,1.5],[2,3]
        kwargs['split'] = [None,None,None,None,3,6]
        kwargs['width'], kwargs['step'] = width, step # take power in 358 samples (700ms) windows width every 51 samples (100ms)

        # Load file :
        loadname = path.join(st.path, 'database/'+rep+'/', fi)
        mat = np.load(loadname,allow_pickle=True)
        print('all files in archive', mat.files)
        x, sf = mat['x'], 512
        n_elec, n_pts, n_trials = x.shape
        print ('--> compute power on : ', rep +'/'+ fi,'shape', x.shape, 'baseline:', baseline)

        # Compute and save power objects :
        powObj_x = power(sf, n_pts,**kwargs, method='hilbert')
        win_all, time = _manageWindow(x.shape[1], width=width, step=step)
        kwargs['time'] = np.array(time) / sf
        kwargs['fname'] = fname
        kwargs['labels'], kwargs['channels']= mat['Mai_RL'], mat['channels']
        kwargs['xyz'] = mat['xyz']
        kwargs['xpow'],  kwargs['xpow_pval']= powObj_x.get(x,n_jobs=-1)
        save_x = path2save+ fi.replace('bipo_all_noWM_physFT.npz', 'bipo_all_'+code+'_6freqs_hilbert=ok.npz')
        np.savez(save_x, **kwargs)
        del kwargs['xpow'],  kwargs['xpow_pval'], kwargs['fname']
        del kwargs, x, sf, n_elec, n_trials
```

### Compute early_late pow files by odor

```python
def early_late_pow(pow_data):
    thr = int(pow_data.shape[-1]/2)
    if thr < 1:
        early = pow_data
        late = np.array([])
    else:
        early = pow_data[...,:thr]
        late = pow_data[...,thr:]
    return early, late
```

```python
from utils import odor_su_score
exp = 'E2' #E:Encoding, R:Retrieval, E2:Reinstatement
st = study('Olfacto')
PATH = join(st.path, 'feature/TPSim_power_data/Power_all_elecs_E_R_by_odor_all_elecsFT/')
filename = join(PATH, '{}_odor_{}_bipo_all_{}_6freqs.npz')
savename = join(PATH, '{}_odor_{}_bipo_all_{}_6freqs_EL.npz')

for su in odor_su_score:
    for od in odor_su_score[su]:
        mat = np.load(filename.format(su,od,exp), allow_pickle=True)
        pow_E, pow_L = early_late_pow(mat['xpow'])
        dict_EL = {}
        for fi in mat.files:
            dict_EL[fi] = mat[fi]
        dict_EL['xpow_E'] = pow_E
        dict_EL['xpow_L'] = pow_L
        print(od,su,'early',pow_E.shape,'late',pow_L.shape)
        
        np.savez(savename.format(su,od,exp),**dict_EL)
```

#### Concatenate early and late pow files according to EM richness

```python
import numpy as np
from brainpipe.system import study
from utils import odor_groups_3wgth as dict_
from os.path import join, exists
from os import makedirs
from itertools import product

st = study('Olfacto')
phases = ['E2']
PATH = join(st.path, 'feature/TPSim_power_data/')
PATH_OD = join(PATH, 'Power_all_elecs_E_R_by_odor_all_elecsFT/')
od_name = join(PATH_OD, '{}_odor_{}_bipo_all_{}_6freqs_EL.npz')
PATH_COND = join(PATH, 'Power_all_elecs_E_R_by_cond_3groups_v=1_elecs=all/')
save_name = join(PATH_COND, '{}_odor_{}_{}_6freqs_EL.npz')

if not exists(PATH_COND):
    makedirs(PATH_COND)
    
for su,phase in product(dict_,phases):
    for cond in dict_[su]:
        print('>>> processing:', su, phase, cond)
        pow_dataE, pow_dataL = np.array([]), np.array([])
        for od in dict_[su][cond]:
            mat = np.load(od_name.format(su,od,phase),allow_pickle=True)
            pow_dataE = np.concatenate((pow_dataE, mat['xpow_E']),axis=-1) \
                                                if np.size(pow_dataE) else mat['xpow_E']
            if mat['xpow_L'].shape[-1] > 0:
                pow_dataL = np.concatenate((pow_dataL, mat['xpow_L']),axis=-1) \
                                                    if np.size(pow_dataL) else mat['xpow_L']
        dict_pow = {}
        for file in mat.files:
            dict_pow[file] = mat[file]
        dict_pow['xpow_E'] = pow_dataE
        dict_pow['xpow_L'] = pow_dataL
        print(su,cond,phase,dict_pow['xpow_E'].shape, dict_pow['xpow_L'].shape)
        np.savez(save_name.format(su,cond,phase),**dict_pow)
```
