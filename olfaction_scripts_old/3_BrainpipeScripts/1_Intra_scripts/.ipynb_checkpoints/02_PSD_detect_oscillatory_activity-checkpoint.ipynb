### Detect and quantify low oscillatory activities

```python
st = study('Ripples')
PATH = join(st.path, 'database/R_odors/')
filename = join(PATH, '{}_cond=ALL_odors=ALL_bipo_INFO_clean.npz')
subjs = ['PIRJ','CHAF','SEMC','FERJ','LEFC','VACJ']
sf = 500
rois = ['HC','orbital','olf']

for roi in rois:
    print(roi)
    psd_roi = np.array([])
    for su in subjs:
        mat = np.load(filename.format(su),allow_pickle=True)
        labels = np.array(['orbital' if lab == 'orbital2' else lab for lab in mat['new_label']])
        if roi in labels:
            data = mat['x'][np.where(labels==roi)[0],...]
            data_sel = data.swapaxes(0,1)
            data_c = np.concatenate(data_sel,axis=-1)
            nelecs,npts = data_c.shape
            for e in range(nelecs):
                freqs, psd = compute_spectrum(data_c[e],sf,
                                    method='welch', avg_type='median',nperseg=sf*2)
                psd_e = psd[np.newaxis]
                psd_roi = np.concatenate((psd_roi,psd_e),axis=0) \
                                            if np.size(psd_roi) else psd_e
    print(roi, psd_roi.shape)
    
    #plot psd averaged across trials
    fmin,fmax = 1,14
    # Initialize a FOOOF object
    fg = FOOOFGroup(peak_width_limits=[1,6], peak_threshold=1., max_n_peaks=4,
                    verbose=False, aperiodic_mode='knee')
    fg.fit(freqs, psd_roi, [fmin,fmax])
    fg.drop(fg.get_params('r_squared') < 0.8)
    fg.print_results()
    #fg.print_settings(description=True)
    fg.plot()
    fg.save_report(PATH+'FOOOF_all_subjs_{}_concat=trials_f={}_{}_max_peak=2'.format(roi,fmin,fmax))
```

### FOOOF with MNE psd across trials

```python
st = study('Ripples')
PATH = join(st.path, 'database/R_odors/')
filename = join(PATH, '{}_cond=ALL_odors=ALL_bipo_INFO_clean.npz')
PATH_SAVE = join(PATH, 'plots_theta_peaks/')
PATH_FG = join(PATH, 'fg_models_theta/')
savename = join(PATH_SAVE, 'theta_peaks_roi={}_f={}_{}.png')
fgname = 'fg_su=all_roi={}_f=[2-9]'
peaks_name = join(PATH_FG, 'peaks_theta_su=all_roi={}_f=[2-9].npy')
if not exists(PATH_SAVE):
    makedirs(PATH_SAVE)
if not exists(PATH_FG):
    makedirs(PATH_FG)
    
subjs = ['PIRJ','CHAF','SEMC','FERJ','LEFC','VACJ']
sf, fmin, fmax = 500, 1, 14
rois = ['olf','HC','orbital']

for roi in rois:
    all_psds = np.array([]) #compute psds by patient because different nb of trials
    for su in subjs:
        mat = np.load(filename.format(su),allow_pickle=True)
        labels = np.array(['orbital' if lab == 'orbital2' else lab for lab in mat['new_label']])
        to_take = [i for i,lab in enumerate(labels) if lab==roi]

        if len(to_take) > 0 :
            data = mat['x'][to_take,...]
            #n_epochs, n_channels, n_times
            data = data.swapaxes(0,-1).swapaxes(1,2)
            chans = np.ndarray.tolist(mat['channel'][to_take])
            
            info = create_info(chans,sfreq=sf,ch_types='eeg')
            epochs = EpochsArray(data,info,tmin=-3,verbose=False)
            psds, freqs = psd_welch(epochs, fmin=fmin, fmax=fmax, n_jobs=1,
                            tmin=0, tmax=2, n_overlap=150, n_fft=sf*2)
            psds_m = np.mean(psds,axis=0)
            all_psds = np.concatenate((all_psds,psds_m),axis=0) if np.size(all_psds) else psds_m
    print(roi, all_psds.shape,freqs.shape)

    #Initialize a FOOOF object
    fg = FOOOFGroup(peak_width_limits=[1,8], min_peak_height=0.2, max_n_peaks=1,
                    peak_threshold=1.,aperiodic_mode='fixed',verbose=False)
    fg.fit(freqs, all_psds, [fmin,fmax])
    
    #get peaks in theta
    bands = Bands({'theta' : [2,9]})
    thetas = get_band_peak_fg(fg, bands.theta) #[CF, PW, BW]
    np.save(peaks_name.format(roi),thetas)
    th_attributes = ['Center Freq', 'Power', 'bandwidth']
    tot_elecs = thetas.shape[0]
    theta_elecs = np.count_nonzero(~np.isnan(thetas[:,0]))

    #plot peaks properties
    fig, axes = plt.subplots(1, 3,sharey=True)
    for i,feat in enumerate(th_attributes):
        axes[i].hist(thetas[:,i])
        axes[i].set_xlabel(feat)
    axes[0].set_ylabel('#elec count')
    fig.suptitle('Theta oscillations in {} ({}/{} elecs)'.format(roi,
                                                    str(theta_elecs),str(tot_elecs)))
    plt.subplots_adjust()
    plt.savefig(savename.format(roi,fmin,fmax))
    
    #drop null models (no fit has been found)
    fg.drop(fg.null_inds_)
    print('null model', fg.null_inds_)
    # Select groups of electrodes with and without peak in theta
    ind_th = [i for i,mod in enumerate(fg) if len(mod.peak_params)>0]
    print('nb of elec with theta', len(ind_th))
    ind_nan = [i for i,mod in enumerate(fg) if len(mod.peak_params)<1]
    print('nb of elec without theta', len(ind_nan))
    elecs_th = fg.get_group(ind_th)
    elecs_nan = fg.get_group(ind_nan)
    print(all_psds.shape)
    #legs = ['PSD-theta','PSD-no-theta']
    #plot_spectra_shading(freqs, [np.mean(all_psds[ind_th,:],axis=0),
    #                    np.mean(all_psds[ind_nan,:],axis=0)],[2,9],
    #                    log_freqs=False, log_powers=True,labels=legs)
    fg.save(fgname.format(roi), file_path=PATH_FG, append=False, 
            save_results=True, save_settings=True, save_data=True)
    
    # Average across individual models fits, specifying bands and an averaging function
    afm = average_fg(elecs_th, bands, avg_method='median')
    afm_nan = average_fg(elecs_nan, bands, avg_method='median')

    fig, axes = plt.subplots(1,2, figsize=(10,5))
    for ax, peak_plot in zip(axes.flatten(), [afm,afm_nan]):
        peak_plot.plot(add_legend=False, ax=ax)
    fig.suptitle('Across electrode models fit in '+roi)
    plt.tight_layout()
    plt.savefig(savename.format(roi,fmin,fmax).replace('.png','_mod_fit.png'))
    #fg.save_report(PATH+'FOOOF_all_subjs_{}_concat=trials_odor_f={}-{}_win=all'.format(roi,fmin,fmax))
```

### K-means to check for spatial clusters in theta patterns
1- create csv with electrodes information
2- look for spatial patterns

```python
st = study('Ripples')
PATH = join(st.path, 'database/R_odors/')
peak_name = join(PATH, 'fg_models_theta/peaks_theta_su=all_roi={}_f=[2-9].npy')
csv_name = join(PATH, 'fg_models_theta/csv_su=all_roi={}_f=[2-9].csv')
pltname = join(PATH, 'fg_models_theta/nb_elec_by_su_roi={}_f=[2-9].png')
raw_name = join(PATH, '{}_cond=ALL_odors=ALL_bipo_INFO_clean.npz')
rois = ['orbital','HC','olf']
plot_figs = True

for roi in rois:
    xyz_roi, su_roi, channels_roi = np.array([]), [], []
    for su in subjects:
        mat = np.load(raw_name.format(su),allow_pickle=True)
        labels = mat['new_label']
        id_labs = np.where(np.array(labels)==roi)[0]
        xyz = mat['xyz'][id_labs,:]
        if len(id_labs) > 0:
            su_roi.extend([su]*xyz.shape[0])
            channels_roi.extend(mat['channel'][id_labs])
            xyz_roi = np.concatenate((xyz_roi,xyz),axis=0) if np.size(xyz_roi) else xyz
    
    peaks = np.load(peaks_name.format(roi))
    detect_theta = [0 if np.isnan(x) else 1 for x in peaks[:,0]]

    data = np.concatenate((np.array(su_roi)[:,np.newaxis],
                           np.array(channels_roi)[:,np.newaxis],
                           xyz_roi,np.array(detect_theta)[:,np.newaxis],
                           peaks),axis=1)
    df = pd.DataFrame(data,columns=['subjects','channels','x','y','z','theta_detect',
                                   'CF','PW','BW'])
    df['theta_detect'] = df['theta_detect'].astype('int')
    df['x'] = df['x'].astype('float')
    df['y'] = df['y'].astype('float')
    df['z'] = df['z'].astype('float')
    if roi == 'OFC_all':
        df = df.loc[df['z']>-20]
    df.to_csv(csv_name.format(roi),index=False)
    
    if plot_figs == True:
        #plot nb of elecs with theta by subj
        df.groupby(['subjects','theta_detect']).theta_detect.count().unstack().plot.bar()
        plt.title('nb of elec in {} by subject'.format(roi))
        plt.savefig(pltname.format(roi))
        plt.clf()
        plt.close()
    
        fig = plt.figure()
        ax = fig.gca()
        #ax = fig.gca(projection='3d')
        for grp_name, grp_idx in df.groupby('theta_detect').groups.items():
            x = df['x'][grp_idx]
            y = df['y'][grp_idx]
            z = df['z'][grp_idx]
            ax.scatter(x,y,label=grp_name)
            #ax.scatter(x,y,z,label=grp_name)
        if roi in ['HC','olf']:
            addLines(plt.gca(),hLines=[-26],vLines=[0])
        if roi == 'orbital':
            xOFC = np.array([20,-20])
            yOFC = np.array([30,32])
            zOFC = np.array([-16,-16])
            ax.scatter(xOFC, yOFC,color='black',s=100)
            #ax.scatter(xOFC, yOFC,zOFC,color='black',s=100)
        plt.legend(loc='upper left')
        plt.title('theta spatial patterns in '+roi)
        plt.savefig(pltname.format(roi).replace('.png','_2D.png'))
```

```python
from mlxtend.evaluate import permutation_test
st = study('Olfacto')
PATH = join(st.path, 'database/Retrieval_all_Odors/')
PATH2 = join(st.path, 'database/Retrieval_No_Odor/')
filename = join(PATH, '{}_odors=all_elecs=all.npz')
filename2 = join(PATH2, '{}_odor_no_odor_bipo_sel_phys.npz')
PATH_SAVE = join(st.path, 'feature_new/R_theta_psd_odor_no_odor/')
subjs = ['PIRJ','CHAF','SEMC','FERJ','LEFC','VACJ']
sf, fmin, fmax = 512, 1, 14

all_trials = False
bands = Bands({'l_theta' : [2,4],
               'h_theta' : [4,8],
              })
rois= ['HC','OFC','pPirT']

for roi in rois:
    #compute psds for all elecs in roi
    all_psd_od, all_psd_no = np.array([]), np.array([])
    su_list = []
    for su in subjs:
        mat = np.load(filename.format(su),allow_pickle=True)
        labels = mat['Mai_RL']
        labels = np.array([lab if lab != 'Amg' else 'pPirT' for lab in mat['Mai_RL']])
        labels = np.array([lab if lab != 'OFC_olf' else 'OFC' for lab in labels])
        labels = np.array([lab if lab != 'aHC' else 'HC' for lab in labels])
        if roi in labels:
            data = mat['x'][np.where(labels==roi)[0],...]
            chans = np.ndarray.tolist(mat['channels'][np.where(labels==roi)[0]])
            
            mat2 = np.load(filename2.format(su),allow_pickle=True)
            id_chs = [i for i, ch in enumerate(mat2['channels']) if ch in chans]
            chans2 = mat2['channels'][id_chs]
            data2 = mat2['x'][id_chs]
            if data.shape[0] != data2.shape[0]:
                chs_data2 = [i for i,ch in enumerate(chans) if ch in chans2]
                data = data[chs_data2,...]
                chans = np.ndarray.tolist(chans2)
                print('correct chans ', data.shape[0]==data2.shape[0])
            
            su_list.extend([su]*len(chans))
            #n_epochs, n_channels, n_times
            data = data.swapaxes(0,-1).swapaxes(1,2)
            data2 = data2.swapaxes(0,-1).swapaxes(1,2)
            
            info = create_info(chans,sfreq=sf,ch_types='eeg')
            epochs = EpochsArray(data,info,tmin=-3,verbose=False)
            epochs2 = EpochsArray(data2,info,tmin=-3,verbose=False)
            
            psds, freqs = psd_welch(epochs, fmin=fmin, fmax=fmax, n_jobs=1,
                            tmin=0, tmax=2, n_overlap=150, n_fft=sf*2)
            psds2, _ = psd_welch(epochs2, fmin=fmin, fmax=fmax, n_jobs=1,
                            tmin=0, tmax=2, n_overlap=150, n_fft=sf*2)
            psds_m, psds_m2 = np.mean(psds,axis=0), np.mean(psds2,axis=0)
            
            all_psd_od = np.concatenate((all_psd_od,psds_m),axis=0) if np.size(all_psd_od) else psds_m
            all_psd_no = np.concatenate((all_psd_no,psds_m2),axis=0) if np.size(all_psd_no) else psds_m2
    
    psds_glob = np.concatenate((all_psd_od[np.newaxis],
                                all_psd_no[np.newaxis]),axis=0)

    #Initialize a FOOOF object
    fg = FOOOFGroup(peak_width_limits=[1,8], min_peak_height=0.2, # max_n_peaks=2,
                    peak_threshold=1.,aperiodic_mode='fixed',verbose=False)
    fgs = fit_fooof_3d(fg, freqs, psds_glob)

    #extract peaks by freq bands
    all_peaks = np.array([])
    for label, band_def in bands:
        print('>>> processing', roi,label)
        peak_od = get_band_peak_fg(fgs[0], band_def)
        peak_no = get_band_peak_fg(fgs[1], band_def)
        mask = np.any(np.isnan(peak_od), axis=1)
        peak_od = peak_od[~mask] #remove electrodes with no peak
        peak_no = peak_no[~mask] #remove electrodes with no peak
        su_sel = np.array(su_list)[~mask]
        su_list_mask = np.concatenate((su_sel,su_sel))
        peak_no_pow = np.nan_to_num(peak_no[:,1],0)
        print('peaks detected in {} electrodes among {} in total'.format(peak_od.shape[0],
                                                                        psds_glob.shape[1]))
        print('CF od', np.mean(peak_od[:,0]), 'CF no od', np.nanmean(peak_no[:,0]))
        print('Pow od', np.mean(peak_od[:,1]), 'Pow no od', np.mean(peak_no_pow))
        p = permutation_test(peak_od[:,1],peak_no_pow,method='approximate',num_rounds=1000)
        print('perm pvalues all elecs', round(p,3))
        
        peaks_pow = np.concatenate((peak_od[:,1],peak_no_pow),axis=0)
        conds_pow = np.array(['PW_od']*peak_od[:,1].shape[0]+['PW_no']*peak_no_pow.shape[0])
        data_df_res = np.concatenate((peaks_pow[:,np.newaxis],conds_pow[:,np.newaxis],
                                      su_list_mask[:,np.newaxis]),axis=1)
        df_res = pd.DataFrame(data_df_res, columns=['PWs','conds','subj'])
        df_res['PWs'] = df_res['PWs'].astype(float)
        md = smf.mixedlm("PWs ~ conds", df_res, groups=su_list_mask)
        mdf = md.fit()
        print('mixed model', roi, mdf.summary())
        df_res_su = df_res.groupby(['subj','conds'],as_index=False).mean()
        p_su = ttest_rel(df_res_su['PWs'].loc[df_res_su['conds']=='PW_od'],
                             df_res_su['PWs'].loc[df_res_su['conds']=='PW_no'])
        print('grouped by subj', p_su)
        
        if p < 0.05:
            #plot results
            x_lab, width = ['Odor', 'No Odor'], 0.35
            data_plot = [np.mean(peak_od[:,1]),np.mean(peak_no_pow)]
            stds = [(0,0), (round(np.std(peak_od[:,1]),2),round(np.std(peak_no_pow),2))]
            plt.bar(np.arange(len(x_lab)),data_plot,width,align='center',
                                                    yerr=stds)
            plt.ylabel('Power')
            plt.title('peaks in {} / Freq = {} / n={}/{} elecs'.format(roi,
                                           band_def,str(peak_od.shape[0]),str(psds_glob.shape[1])))
            plt.xticks(np.arange(len(x_lab)),labels=x_lab)

            # create empty labels
            handles = [mpl_patches.Rectangle((0, 0), 1, 1, fc="white", ec="white", 
                                     lw=0, alpha=0)] * 3
            # create the corresponding number of labels (= the text you want to display)
            infos = []
            infos.append("p = {0:.3g}".format(p))
            infos.append("CF(odor) = {0:.3g}".format(np.mean(peak_od[:,0])))
            infos.append("CF(No odor) = {0:.3g}".format(np.nanmean(peak_no[:,0])))

            plt.legend(handles, infos, loc='best', fontsize='small', 
                      fancybox=True, framealpha=0.7, handlelength=0, handletextpad=0)
            plt.savefig(PATH_SAVE+'Pow_freq={}_roi={}_f={}.png'.format(label,roi,str(band_def)))
            plt.close()
            plt.clf()
            print('\n')
```

```python
from mlxtend.evaluate import permutation_test
st = study('Olfacto')
PATH = join(st.path, 'database/Encoding_all_Odors/')
filename = join(PATH, '{}_all_odors_all_elecs.npz')
PATH_SAVE = join(st.path, 'feature_new/theta_psd_odor_no_odor/')
filesave = join(PATH, 'all_su_elecs_to_take_roi={}_theta_2_14.npz')
subjs = ['PIRJ','CHAF','SEMC','FERJ','LEFC','VACJ']
sf, fmin, fmax = 512, 1, 14
rois = ['HC', 'IFG', 'Ins_olf', 'MFG', 'OFC', 'OFC_olf',
                 'aHC', 'pPirT', 'temp']

all_trials = False
bands = Bands({'l_theta' : [2,6],
               'h_theta' : [6,14],
               #'ext_theta': [6,9]
              })

for roi in rois:
    #compute psds for all elecs in roi
    all_psd_od, all_psd_no = np.array([]), np.array([])
    all_subjs, all_chs = np.array([]), np.array([])
    for su in subjs:
        mat = np.load(filename.format(su),allow_pickle=True)
        labels = mat['Mai_RL']
        labels = np.array([lab if lab != 'Amg' else 'pPirT' for lab in mat['Mai_RL']])
        if roi in labels:
            data = mat['x'][np.where(labels==roi)[0],...]
            chans = np.ndarray.tolist(mat['channels'][np.where(labels==roi)[0]])
            
            #concat infos into a big file by ROI
            all_subjs = np.concatenate((all_subjs,[su]*len(chans))) if np.size(all_subjs) else [su]*len(chans)
            all_chs = np.concatenate((all_chs,chans)) if np.size(all_chs) else chans
            
            #n_epochs, n_channels, n_times
            data = data.swapaxes(0,-1).swapaxes(1,2)
            info = create_info(chans,sfreq=sf,ch_types='eeg')
            epochs = EpochsArray(data,info,tmin=-3,verbose=False)
            
            psds, freqs = psd_welch(epochs, fmin=fmin, fmax=fmax, n_jobs=1,
                            tmin=0, tmax=2, n_overlap=150, n_fft=sf*2)
            psds2, _ = psd_welch(epochs, fmin=fmin, fmax=fmax, n_jobs=1,
                            tmin=-2.5, tmax=-0.5, n_overlap=150, n_fft=sf*2)
            psds_m, psds_m2 = np.mean(psds,axis=0), np.mean(psds2,axis=0)
            
            all_psd_od = np.concatenate((all_psd_od,psds_m),axis=0) if np.size(all_psd_od) else psds_m
            all_psd_no = np.concatenate((all_psd_no,psds_m2),axis=0) if np.size(all_psd_no) else psds_m2
    
    psds_glob = np.concatenate((all_psd_od[np.newaxis],
                                all_psd_no[np.newaxis]),axis=0)

    #Initialize a FOOOF object
    fg = FOOOFGroup(peak_width_limits=[1,8], min_peak_height=0.2, # max_n_peaks=2,
                    peak_threshold=1.,aperiodic_mode='fixed',verbose=False)
    fgs = fit_fooof_3d(fg, freqs, psds_glob)

    #extract peaks by freq bands
    all_peaks, all_masks = np.array([]), []
    for label, band_def in bands:
        print('>>> processing', roi,label)
        peak_od = get_band_peak_fg(fgs[0], band_def)
        peak_no = get_band_peak_fg(fgs[1], band_def)
        nan_indx = np.ravel(np.argwhere(~np.isnan(peak_od[:,0])))
        all_masks.append(nan_indx)
        mask = np.any(np.isnan(peak_od), axis=1)
        peak_od = peak_od[~mask] #remove electrodes with no peak
        peak_no = peak_no[~mask] #remove electrodes with no peak
        peak_no_pow = np.nan_to_num(peak_no[:,1],0)
        print('peaks detected in {} electrodes among {} in total'.format(peak_od.shape[0],
                                                                        psds_glob.shape[1]))
        print('CF od', np.mean(peak_od[:,0]), 'CF no od', np.nanmean(peak_no[:,0]))
        print('Pow od', np.mean(peak_od[:,1]), 'Pow no od', np.mean(peak_no_pow))
        #p = permutation_test(peak_od[:,1],peak_no_pow,method='approximate',num_rounds=1000)
        T,p = mannwhitneyu(peak_od[:,1],peak_no_pow,alternative='greater')
        print('pvalues ', round(p,3))
        
        #plot results
        x_lab, width = ['POST', 'PRE'], 0.35
        data_plot = [np.mean(peak_od[:,1]),np.mean(peak_no_pow)]
        stds = [(0,0), (round(np.std(peak_od[:,1]),2),round(np.std(peak_no_pow),2))]
        plt.bar(np.arange(len(x_lab)),data_plot,width,align='center',
                                                yerr=stds)
        plt.ylabel('Power')
        plt.title('peaks in {} / Freq = {} / n={}/{} elecs'.format(roi,
                                       band_def,str(peak_od.shape[0]),str(psds_glob.shape[1])))
        plt.xticks(np.arange(len(x_lab)),labels=x_lab)
        
        # create empty labels
        handles = [mpl_patches.Rectangle((0, 0), 1, 1, fc="white", ec="white", 
                                 lw=0, alpha=0)] * 3
        # create the corresponding number of labels (= the text you want to display)
        infos = []rois = ['HC', 'IFG', 'Ins_olf', 'MFG', 'OFC', 'OFC_olf',
                 'aHC', 'pPirT', 'temp']
        infos.append("p = {0:.3g}".format(p))
        infos.append("CF(POST) = {0:.3g}".format(np.mean(peak_od[:,0])))
        infos.append("CF(PRE) = {0:.3g}".format(np.nanmean(peak_no[:,0])))

        plt.legend(handles, infos, loc='best', fontsize='small', 
                  fancybox=True, framealpha=0.7, handlelength=0, handletextpad=0)
        plt.savefig(PATH_SAVE+'Pow_freq={}_roi={}_f={}_pre_post.png'.format(label,roi,str(band_def)))
        plt.close()
        plt.clf()
        print('\n')
    dict_save = {}
    dict_save['channels'] = all_chs
    dict_save['subjects'] = all_subjs
    dict_save['freqs'] = np.asarray([f for f in bands])
    for i, (label, band_def) in enumerate(bands):
        dict_save['to_take_'+label] = all_masks[i]
    print(all_chs.shape,all_subjs.shape,roi)
    np.savez(filesave.format(roi),**dict_save)
```
