```python
from os.path import isfile, join, exists
from os import listdir, makedirs
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.offsetbox import AnchoredText
from brainpipe.system import study
from scipy import stats
from itertools import product
```

### Correlations BY ODOR // Mean power in specific roi and memory scores

```python
feat, corr = 'pow', False
###############################################################################
st = study('Olfacto')
path_npz = join(st.path, 'feature/0_Power_Encoding_EpiPerf_By_Odor/')
path2save = join(st.path, 'correlations_pow_norm/encoding/')
npz_form = join(path_npz, '{}_odor_{}_bipo_sel_physFT_pow.npz')
f_form_save = join(path2save, 'Correl_{}_{}_{}_{}.png')
###############################################################################

su_codes = ['S0','S1','S2','S3','S4','S5','S6']
freqs = ['0_VLFC', '1_delta', '2_theta', '3_alpha', '4_beta','5_gamma1','6_gamma2']
method = ['s_Mail_RL']
wins = [5,10,15,20,25,30]
times = ['-500-0ms','0-500ms','500-1000ms','1000-1500ms','1500-2000ms']

odors_su = {'CHAF': {5:12,7:68,8:36,9:96,1:6,2:2,3:68,4:8},
            'LEFC': {1:4,2:0,3:6,4:12,14:96,15:2,16:4,17:68},
            'PIRJ': {4:36,9:2,1:4,18:32,6:34,5:4,7:68}, #missing odor 5
            'VACJ': {14:6,15:64,16:68,17:8,10:6,11:4,12:4,13:40},
            'SEMC': {10:2,11:6,12:6,13:6,5:8,7:4,8:8,9:10},
            'MICP': {2:6,12:8,6:96,8:8,3:12,18:4,9:6,14:10},
            'FERJ': {16:6,17:6,5:8,7:6,12:8,13:8,2:6,1:10}}

rois_list = {'Frontal':['ACC','IFG','MFG','OFC','SFG','Ins'],
            'Olf':['Amg','pPirT','Amg-PirT','Ins','OFC'],
            'MTL':['HC','PHG'],
            'Temporal':['FuG','HC','PHG','ITG','MTG','STG','Amg','pPirT','Amg-PirT'],
            'all':['ACC','IFG','MFG','OFC','SFG','Ins','FuG','HC','PHG','ITG','MTG','STG','Amg','pPirT','Amg-PirT']}

for i,freq in enumerate(freqs):
    for roi in rois:
        for w,win in enumerate(wins[:-1]):
            pow_odors, scores_odors = [], []
            for su in odors_su:
                for odor in odors_su[su]:
                    #print('su',su,'in',roi,'for odor',odor,'score',odors_su[su][odor])
                    arch_sig = np.load(npz_form.format(su,odor))
                    Mai_RL= arch_sig['Mai_RL']
                    power = arch_sig['xpow'][i,:,win:wins[w+1],:] #nfreq,nelecs,npts,ntrials
                    pow_odors = np.append(pow_odors,np.mean(power[np.where(Mai_RL==roi)]))
                    scores_odors = np.append(scores_odors,odors_su[su][odor])
            scores_odors = np.array(scores_odors)[np.where(~np.isnan(pow_odors))]
            pow_odors = pow_odors[np.logical_not(np.isnan(pow_odors))]
            print(roi,len(scores_odors),len(pow_odors))
            #Plot all correlations
            R, p = stats.pearsonr(scores_odors,pow_odors)
            #R, p = round(R,2), round(p,3)
            print('stats for',freq,roi,R,p)
            if p <= 0.05:
                fig, ax = plt.subplots()
                plt.title('Correlation btw Epi score and '+freq+' power in '+roi+' win '+times[w])
                plt.xlabel('EpiScore'), plt.ylabel(freq)
                anchored_text = AnchoredText('R coeff = %s, pval = %s' % (R, p), loc=2)
                ax.add_artist(anchored_text)
                fit = np.polyfit(scores_odors, pow_odors, deg=1)
                ax.plot(sorted(scores_odors), fit[0] * np.array(sorted(scores_odors)) + fit[1], color='red')
                ax.scatter(scores_odors, pow_odors, color = 'dodgerblue')
                plt.savefig(f_form_save.format(freq,roi,'EpiScore',times[w]))
                plt.clf()
                plt.close()

for i,freq in enumerate(freqs):    
    for w,win in enumerate(wins[:-1]):
        pow_odors, scores_odors = [], []
        for s,su in enumerate(odors_su):
            for odor in odors_su[su]:
                #print('su',su,'in',roi,'for odor',odor,'score',odors_su[su][odor])
                arch_sig = np.load(npz_form.format(su,odor))
                Mai_RL= arch_sig['Mai_RL']
                power = arch_sig['xpow'][i,:,win:wins[w+1],:] #nfreq,nelecs,npts,ntrials
                pow_odors = np.append(pow_odors,np.mean(power[np.where((Mai_RL==rois_PC[0])|(Mai_RL==rois_PC[1])|(Mai_RL==rois_PC[2]))]))
                scores_odors = np.append(scores_odors,odors_su[su][odor])
        scores_odors = np.array(scores_odors)[np.where(~np.isnan(pow_odors))]
        pow_odors = pow_odors[np.logical_not(np.isnan(pow_odors))]
        print(roi,len(scores_odors),len(pow_odors))
        #Plot all correlations
        R, p = stats.pearsonr(scores_odors,pow_odors)
        #R, p = round(R,2), round(p,3)
        print('stats for',freq,'olf PC',R,p)
        if p <= 0.05:
            fig, ax = plt.subplots()
            plt.title('Correlation btw Epi score and '+freq+' power in '+roi+' win '+times[w])
            plt.xlabel('EpiScore'), plt.ylabel(freq)
            anchored_text = AnchoredText('R coeff = %s, pval = %s' % (R, p), loc=2)
            ax.add_artist(anchored_text)
            fit = np.polyfit(scores_odors, pow_odors, deg=1)
            ax.plot(sorted(scores_odors), fit[0] * np.array(sorted(scores_odors)) + fit[1], color='red')
            ax.scatter(scores_odors, pow_odors, color = 'dodgerblue')
            plt.savefig(f_form_save.format(freq,'olf','EpiScore',times[w]))
            plt.clf()
            plt.close()
```

### Correlations BY PATIENT // Mean power in a specific roi and memory scores

```python
feat, corr = 'pow', False
###############################################################################
st = study('Olfacto')
path_npz = join(st.path, 'feature/0_Power_Encoding_EpiPerf_By_Odor/')
path2save = join(st.path, 'correlations_pow/by_odor/')
npz_form = join(path_npz, '{}_odor_{}_bipo_sel_physFT_pow.npz')
f_form_save = join(path2save, 'Correl_{}_{}_{}_{}_bsl2.png')
###############################################################################

subjects = ['CHAF','FERJ','LEFC','SEMC','PIRJ','MICP','VACJ']
su_codes = ['S0','S1','S2','S3','S4','S5','S6']
freqs = ['0_VLFC', '1_delta', '2_theta', '3_alpha', '4_beta','5_gamma1','6_gamma2']
method = ['s_Mail_RL']

# Behavioral Performance
scores_tot = [12.3,2.4,8.0,2.1,7.7,6.3,8.3]
scores_low = [2.3,2.0,0.8,1.6,1.2,1.8,1.7]
scores_high = [22.3,2.8,15.2,2.9,14.2,8.9,15.5]
scores_diff = [scores_high[i]-scores_low[i] for i in range(len(scores_high))]

for freq in freqs:
    # Load elecs infos and power results
    arch_sig = np.load(npz_form.format('All_subjects',freq, 'odor'))
    su_codes, s_Mai_RL= arch_sig['su_codes'], arch_sig['s_MAI_RL']
    rois_list = np.unique(s_Mai_RL)
    rois_olf = ['Amg','pPirT','Amg-PirT']
    rois_no_olf = [roi for roi in rois_list if roi not in rois_olf]
    s_power = ((arch_sig['s_elec_pow1'][:,5:30]-arch_sig['s_elec_pow0'][:,5:30])/arch_sig['s_elec_pow0'][:,5:30])*100
    
    # Select only electrodes in specific rois
    for roi in rois_no_olf:
        #Create an array of mean power by subject and roi
        pow_by_su = np.array([])
        for s,su in enumerate(subjects):
            pow_su = np.mean(s_power[np.where((su_codes=='S'+str(s))&(s_Mai_RL==roi))])
            pow_by_su = np.hstack((pow_by_su,pow_su)) if np.size(pow_by_su) else pow_su
        #remove scores values when patient not included
        scores_diff_sel = np.array(scores_diff)[np.where(~np.isnan(pow_by_su))]
        scores_tot_sel = np.array(scores_tot)[np.where(~np.isnan(pow_by_su))]
        pow_by_su = pow_by_su[np.logical_not(np.isnan(pow_by_su))]
        print(roi, pow_by_su.shape)
        print('scores',scores_diff_sel, scores_tot_sel)

        #Plot all correlations
        scores, scores_names = [scores_tot_sel,scores_diff_sel], ['tot_EM','diff_EM',]
        if pow_by_su.shape[0] > 5:
            for i,score in enumerate(scores):
                R, p = stats.pearsonr(score,pow_by_su)
                R, p = round(R,2), round(p,3)
                print('stats for',freq,scores_names[i],R,p)
                if p <= 0.05:
                    fig, ax = plt.subplots()
                    plt.title('Correlation btw '+scores_names[i]+' and '+' for '+freq+' in'+roi)
                    plt.xlabel(scores_names[i]), plt.ylabel(freq)
                    anchored_text = AnchoredText('R coeff = %s, pval = %s' % (R, p), loc=2)
                    ax.add_artist(anchored_text)
                    fit = np.polyfit(score, pow_by_su, deg=1)
                    ax.plot(sorted(score), fit[0] * np.array(sorted(score)) + fit[1], color='red')
                    ax.scatter(score, pow_by_su, color = 'dodgerblue')
                    plt.savefig(f_form_save.format(freq,roi,scores_names[i]))
                    plt.clf()
                    plt.close()
        
for freq in freqs:    
    pow_by_su = np.array([])
    for s,su in enumerate(subjects):
        pow_su = np.mean(s_power[np.where((su_codes=='S'+str(s))&((s_Mai_RL==rois_olf[0])|(s_Mai_RL==rois_olf[1])|(s_Mai_RL==rois_olf[2])))])
        pow_by_su = np.hstack((pow_by_su,pow_su)) if np.size(pow_by_su) else pow_su
    #remove scores values when patient not included
    scores_diff_sel = np.array(scores_diff)[np.where(~np.isnan(pow_by_su))]
    scores_tot_sel = np.array(scores_tot)[np.where(~np.isnan(pow_by_su))]
    pow_by_su = pow_by_su[np.logical_not(np.isnan(pow_by_su))]
    print('olf', pow_by_su.shape)
    print('scores',scores_diff_sel, scores_tot_sel)

    #Plot all correlations
    scores, scores_names = [scores_tot_sel,scores_diff_sel], ['tot_EM','diff_EM',]
    if pow_by_su.shape[0] > 5:
        for i,score in enumerate(scores):
            R, p = stats.pearsonr(score,pow_by_su)
            R, p = round(R,2), round(p,3)
            print('stats for',freq,scores_names[i],R,p)
            if p <= 0.05:
                fig, ax = plt.subplots()
                plt.title('Correlation btw '+scores_names[i]+' and '+' for '+freq+' in'+roi)
                plt.xlabel(scores_names[i]), plt.ylabel(freq)
                anchored_text = AnchoredText('R coeff = %s, pval = %s' % (R, p), loc=2)
                ax.add_artist(anchored_text)
                fit = np.polyfit(score, pow_by_su, deg=1)
                ax.plot(sorted(score), fit[0] * np.array(sorted(score)) + fit[1], color='red')
                ax.scatter(score, pow_by_su, color = 'dodgerblue')
                plt.savefig(f_form_save.format(freq,roi,scores_names[i]))
                plt.clf()
                plt.close()
            
```

```python
feat, corr = 'pow', False
###############################################################################
st = study('Olfacto')
path_npz = join(st.path, 'feature/0_Power_Retrieval_EpiPerf_LowHigh/')
path2save = join(st.path, 'correlations_pow/retrieval/')
npz_form = join(path_npz, '{}_odor_{}_bipo_sel_physFT_pow.npz')
f_form_save = join(path2save, 'Correl_{}_{}_{}.png')
if not exists(path2save):
    makedirs(path2save)
############################################################################### 

subjects = ['CHAF','FERJ','LEFC','SEMC','PIRJ','MICP','VACJ']
su_codes = ['S0','S1','S2','S3','S4','S5','S6']
freqs = ['0_VLFC', '1_delta', '2_theta', '3_alpha', '4_beta','5_gamma1','6_gamma2']
wins = [5,10,15,20,25,30]
times = ['-500-0ms','0-500ms','500-1000ms','1000-1500ms','1500-2000ms'] #['-500-0ms','0-1000ms','1000-2000ms']

odors_su = {'CHAF': {5:12,7:68,8:36,9:96,1:6,2:2,3:68,4:8},
            'LEFC': {1:4,2:0,3:6,4:12,14:96,15:2,16:4,17:68},
            'PIRJ': {4:36,9:2,1:4,18:32,6:34,5:4,7:68}, #missing odor 5
            'VACJ': {14:6,15:64,16:68,17:8,10:6,11:4,12:4,13:40},
            'SEMC': {10:2,11:6,12:6,13:6,5:8,7:4,8:8,9:10},
            'MICP': {2:6,12:8,6:96,8:8,3:12,18:4,9:6,14:10},
            'FERJ': {16:6,17:6,5:8,7:6,12:8,13:8,2:6,1:10}}

rois_list = {'Frontal':['ACC','IFG','MFG','OFC','SFG','Ins'],
            'Olf':['Amg','pPirT','Amg-PirT','Ins','OFC'],
            'MTL':['HC','PHG'],
            'Temporal':['FuG','HC','PHG','ITG','MTG','STG','Amg','pPirT','Amg-PirT'],
            'all':['ACC','IFG','MFG','OFC','SFG','Ins','FuG','HC','PHG','ITG','MTG','STG','Amg','pPirT','Amg-PirT']}

for i,freq in enumerate(freqs):
    for region in rois_list:
        for w,win in enumerate(wins[:-1]):
            pow_norm, scores_norm = [], []
            for s,su in enumerate(odors_su):
                pow_rois, score_rois = [], []
                for roi,odor in product(rois_list[region],odors_su[su]):
                    #print('su',su,'in',roi,'for odor',odor,'score',odors_su[su][odor])
                    arch_sig = np.load(npz_form.format(su,odor))
                    Mai_RL= arch_sig['Mai_RL']
                    power = np.mean(arch_sig['xpow'][i,:,win:wins[w+1],:],axis=(1,2))#nfreq,nelecs,npts,ntrials
                    #print('power',power.shape)
                    power_roi = power[np.where(Mai_RL==roi)]
                    pow_rois = np.append(pow_rois, power_roi)
                    score_rois= np.append(score_rois,[odors_su[su][odor]]*power_roi.shape[0])
                    #print('score, power',power_roi, pow_rois, score_rois)
            pow_su = (pow_rois - pow_rois.mean())/pow_rois.std()
            score_su = (score_rois - score_rois.mean())/score_rois.std()
            pow_norm = np.append(pow_norm,pow_su)
            scores_norm = np.append(scores_norm,score_su)
            print(freq,region,'check',pow_norm.shape,scores_norm.shape)
            scores_norm = np.array(scores_norm)[np.where(~np.isnan(pow_norm))]
            pow_norm = pow_norm[np.logical_not(np.isnan(pow_norm))]
            print(region,len(scores_norm),len(pow_norm))
            #Plot all correlations
            R, p = stats.pearsonr(scores_norm,pow_norm)
            #R, p = round(R,2), round(p,3)
            print('stats for',region,freq,R,p)
            if p <= 0.05:
                fig, ax = plt.subplots()
                plt.title('Correlation btw Epi score and '+freq+' power in '+region+'for '+times[w])
                plt.xlabel('EpiScore'), plt.ylabel(freq)
                anchored_text = AnchoredText('R coeff = %s, pval = %s' % (R, p), loc=2)
                ax.add_artist(anchored_text)
                fit = np.polyfit(scores_norm, pow_norm, deg=1)
                ax.plot(sorted(scores_norm), fit[0] * np.array(sorted(scores_norm)) + fit[1], color='red')
                ax.scatter(scores_norm, pow_norm, color = 'dodgerblue')
                plt.savefig(f_form_save.format(freq,region,'EpiScore',times[w]))
                plt.clf()
                plt.close()
    
```
