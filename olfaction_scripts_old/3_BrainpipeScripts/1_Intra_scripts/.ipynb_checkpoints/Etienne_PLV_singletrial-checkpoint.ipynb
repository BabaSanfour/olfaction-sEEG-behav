## Function for computing the PLV

```python
import numpy as np
from scipy import signal
import xarray as xr

from mne.time_frequency import morlet
from mne.utils import ProgressBar

from frites.io import set_log_level, logger
from frites.dataset import SubjectEphy


def conn_plv_spectrum(data, sfreq, freqs, n_cycles=7., roi=None, times=None,
                      poi=None, verbose=None):
    """Compute the PLV spectrum.
    This function perform the following steps :
        * Get the complex decomposition in a specific frequency band using
          Morlet's wavelets
        * Compute the PLV across time points
    Parameters
    ----------
    data : array_like
        Electrophysiological data. Several input types are supported :
            * Standard NumPy arrays of shape (n_epochs, n_roi, n_times)
            * mne.Epochs
            * xarray.DataArray of shape (n_epochs, n_roi, n_times)
    sfreq : float
        The sampling frequency
    freqs : array_like
        Array of frequencies
    n_cycles : array_like | 7.
        Number of cycles to use for each frequency. If a float or an integer is
        used, the same number of cycles is going to be used for all frequencies
    times : array_like | None
        Time vector array of shape (n_times,). If the input is an xarray, the
        name of the time dimension can be provided
    roi : array_like | None
        ROI names of a single subject. If the input is an xarray, the
        name of the ROI dimension can be provided
    poi : tuple | None
        Period of interest in time units. Once the complex decomposition is
        performed on the entire time series, you can select a smaller temporal
        period of interest (e.g (0, 1.5) to select the period between 0s and
        1.5s)

    Returns
    -------
    plv : xr.DataArray
        PLV array of shape (n_trials, n_pairs, n_freqs)
    """
    set_log_level(verbose)

    # ___________________________________ I/O _________________________________
    if isinstance(data, xr.DataArray):
        trials, attrs = data[data.dims[0]].data, data.attrs
    else:
        trials, attrs = np.arange(data.shape[0]), {}
    # internal conversion
    data = SubjectEphy(data, y=trials, roi=roi, times=times)
    x, roi, times = data.data, data['roi'].data, data['times'].data
    trials = data['y'].data
    n_trials, n_roi, n_times = data.shape
    # inputs checking
    assert isinstance(freqs, (np.ndarray, list, tuple))
    if isinstance(n_cycles, (int, float)):
        n_cycles = [n_cycles] * len(freqs)
    assert len(n_cycles) == len(freqs)
    freqs, n_cycles = np.array(freqs), np.array(n_cycles)
    n_freqs = len(freqs)
    # period of interest
    poi = (times[0], times[-1]) if poi is None else poi
    idx_poi = np.abs(times.reshape(-1, 1) - np.array(
        poi).reshape(1, -1)).argmin(0)
    # get the non-directed pairs and build roi pairs names
    x_s, x_t = np.triu_indices(n_roi, k=1)
    roi_p = [f"{roi[s]}-{roi[t]}" for s, t in zip(x_s, x_t)]
    n_pairs = len(roi_p)

    # ___________________________________ PLV _________________________________
    logger.info(f"Compute pairwise PLV spectrum (n_pairs={n_pairs}, "
                f"n_freqs={n_freqs}), poi={idx_poi}")
    pbar = ProgressBar(range(n_freqs), mesg='PLV spectrum estimation')
    # move data to the gpu
    x = np.asarray(x)
    # compute plv for each trial
    plv = np.zeros((n_trials, n_pairs, n_freqs))
    for n_f in range(n_freqs):
        # get the wavelet
        wav = np.asarray(morlet(sfreq, freqs[[n_f]], n_cycles=n_cycles[[n_f]]))
        wav = wav.reshape(1, 1, -1)
        # compute the complex decomposition
        dcomp = signal.fftconvolve(x, wav, mode='same', axes=2)
        # poi selection
        dcomp = dcomp[..., idx_poi[0]:idx_poi[-1] + 1]
        # now get the phase
        pha = np.angle(dcomp)
        for n_p, (s, t) in enumerate(zip(x_s, x_t)):
            da_s = pha[:, s, :]
            da_t = pha[:, t, :]
            plv[:, n_p, n_f] = np.abs(np.exp(1j * (da_s - da_t)).sum(-1))
        pbar.update_with_increment_value(1)
    # move it from gpu to cpu
    plv = np.asarray(plv / float(n_times))

    # _________________________________ XARRAY ________________________________
    plv = xr.DataArray(plv, dims=('trials', 'roi', 'freqs'),
                       coords=(trials, roi_p, freqs), name='plv_spectrum')
    # set attributes
    cfg = dict(n_cycles=n_cycles, sfreq=sfreq, poi=poi)
    plv.attrs = {**attrs, **cfg}

    return plv
```

## Test the function with simulated data
### Providing NumPy inputs

La fonction ci-dessus peut prendre en charge trois type d'inputs :
- NumPy arrays
- MNE epochs
- Xarray (le format que personnelement j'utilise le plus maintenant pour travailler avec mes données intra / MEG)

```python
###############################################################################
# data settings
n_elec = 3       # number of sEEG contacts
n_epochs = 50    # number of trials / epochs / repetitions
n_times = 1000   # number of time points
sfreq = 256.     # sampling frequency
###############################################################################

# generate random data
data = np.random.rand(n_epochs, n_elec, n_times)
elec = ["elec_%i" % (k + 1) for k in range(n_elec)]
times = np.arange(n_times) / sfreq

# simulate a theta coupling between electrodes 1 and 2 between time points [10, 500]
sl_th = slice(10, 500)
times_th = times[sl_th]
theta = (np.sin(2 * np.pi * 6. * times_th) * np.hanning(sl_th.stop - sl_th.start)).reshape(1, -1)
data[:, 0, sl_th] += theta
data[:, 1, sl_th] += theta

# simulate an alpha coupling between electrodes 2 and 3 between time points [500, 900]
sl_al = slice(500, 900)
times_al = times[sl_al]
alpha = (np.sin(2 * np.pi * 10. * times_al) * np.hanning(sl_al.stop - sl_al.start)).reshape(1, -1)
data[:, 1, sl_al] += alpha
data[:, 2, sl_al] += alpha

# DataArray conversion (for plotting only)
data_xr = xr.DataArray(data, dims=('trials', 'elec', 'times'),
                       coords=(np.arange(n_epochs), elec, times))

# plot the data
data_xr.mean('trials').plot.line(x='times', col='elec');
plt.gcf().suptitle('Mean raw activity per electrode', y=1.02);
```

On voit bien que les électrodes 1 et deux ont une synchronisation en theta entre ~[0.25, 1.75]s et en alpha pour les électrodes 2 et 3 entre ~[2.25, 3.5]s

```python
###############################################################################
# plv settings
f_start = 2.   # starting frequency
f_end = 50.    # ending frequency
n_freqs = 100  # number of frequencies
freqs = np.linspace(f_start, f_end, n_freqs)
###############################################################################

plv = conn_plv_spectrum(
    data, sfreq, freqs, times=times, roi=elec
)

# plot the result
plv.mean('trials').plot.line(x='freqs', col='roi');

"""
on retrouve le couplage entre les électrodes 1 et 2 en theta (~6hz) et
entre 2 et 3 en alpha (~10hz)
"""
```
