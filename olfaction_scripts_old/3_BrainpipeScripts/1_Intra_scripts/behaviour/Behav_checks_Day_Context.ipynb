```python
import numpy as np
from os.path import join
import pandas as pd
from utils import odor_groups_3wgth
import pingouin as pg
from scipy.stats import ttest_rel

PATH = '/media/1_Analyses_Intra_EM_Odor/1bis_OE_BaseSam/JPlailly201306_seeg_ALS/'
dfname = join(PATH, 'behavior/episodic_all_su_all_trials_old_odors.csv')


#invert dict by odor instead of by condition
new_dict = {}
for su in odor_groups_3wgth:
    all_odors = {}
    for k, v in odor_groups_3wgth[su].items():
        all_odors.update({od:k for od in v})
    new_dict[su] = all_odors

df = pd.read_csv(dfname,index_col=0)
print(df.shape)

#clean df and column type
df['odor_num'] = df['odor_num'].astype('str')
df['image_context'] = df['image_context'].astype('str')
df['context_day'] = df['context_day'].astype('str')
df = df.drop(df[(df['odor_num'] == '15') & (df['subject_name'] == 'PIRJ')].index)

df_gr = df.groupby(['subject_name','image_context'], as_index=False).mean()
a = df_gr['epi_chance_wght'].loc[df_gr['image_context']=="M"].values
b = df_gr['epi_chance_wght'].loc[df_gr['image_context']=="F"].values
T, p = ttest_rel(a,b)
print('context', np.mean(a), np.mean(b), T,p)

df_gr = df.groupby(['subject_name','context_day'], as_index=False).mean()
a = df_gr['epi_chance_wght'].loc[df_gr['context_day']=="1.0"].values
b = df_gr['epi_chance_wght'].loc[df_gr['context_day']=="2.0"].values
T, p = ttest_rel(a,b)
print('day',np.mean(a), np.mean(b), T,p)
#anova 2 facteurs à mesures répétées (pas assez d'essais)
#df_sel = df.groupby(['subject_name','odor_num'],as_index=False).first()
#df_sel['richness'] = [new_dict[su][od] for su, od in zip(df_sel['subject_name'],
#                                                         df_sel['odor_num'])]
#df_sel['idx'] = [1 for w in df_sel['richness']]
#df_im = df_sel.groupby(['subject_name','image_context','richness'],as_index=False).sum()
#aov1 = pg.rm_anova(dv='idx',within=['image_context','richness'],
#                  subject='subject_name',data=df_im)
```
