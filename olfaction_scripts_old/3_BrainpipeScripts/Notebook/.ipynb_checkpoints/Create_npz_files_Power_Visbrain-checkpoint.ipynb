## Npz files for Visbrain for all subjects

```python
from brainpipe.system import study
import numpy as np
from os import listdir
from os.path import isfile, join
import numpy as np

st = study('Olfacto')
bsl = 'None'
phase = 'odor'
path_classif = join(st.path, 'classified/2_Classif_Power_time_allrois_RecScore_win700_step100_expi/')
path_pow = join(st.path, 'feature/7_Power_E1E2_Odor_Good_Bad_RecScore_Expi/')
path2save = join(st.path, 'Visbrain/npz_files_odor_expi/')
subjects = ['LEFC','CHAF','VACJ','SEMC','FERJ','MICP','PIRJ'] #respectively S0 S1 S2 S3 S4 S5
freqs = ['0_VLFC', '1_delta', '2_theta', '3_alpha', '4_beta','5_gamma1','6_gamma2']

## Create a npz file with all sources informations + POWER (subjects, coords, labels)
for j,freq in enumerate(freqs):
    su_codes, s_xyz, s_channels,s_labels, s_elec, s_elec_pow_bad, s_elec_pow_good = np.array([]),np.array([]), np.array([]), np.array([]),np.array([]),np.array([]),np.array([])
    s_stats, s_stats_t0, s_da, s_th_05, s_th_01 = np.array([]),np.array([]),np.array([]),np.array([]),np.array([])
    s_BA, s_AAL, s_AAL_RL = np.array([]), np.array([]),np.array([])
    print(freq)
    for i,su in enumerate(subjects):
        print(su)
        pow_mat_bad = np.load(path_pow+su+'_odor_bad_bipo_sel_phys_'+str(bsl)+'_500_pow.npz')
        pow_mat_good = np.load(path_pow+su+'_odor_good_bipo_sel_phys_'+str(bsl)+'_500_pow.npz')
        #print(pow_mat_bad.files, pow_mat_good.files)
        xyz_all = pow_mat_bad['xyz']
        for elec in range(xyz_all.shape[0]):
            ### INFO electrodes for all subjects
            su_codes = np.vstack((su_codes,'S'+str(i))) if np.size(su_codes) else 'S'+str(i)
            xyz = xyz_all[elec]
            s_xyz = np.vstack((s_xyz, xyz)) if np.size(s_xyz) else xyz
            aal = pow_mat_bad['s_aal'][elec]
            s_AAL = np.vstack((s_AAL, aal)) if np.size(s_AAL) else aal
            aal_RL = aal[:-4]
            s_AAL_RL = np.vstack((s_AAL_RL, aal_RL)) if np.size(s_AAL_RL) else aal_RL
            BA = pow_mat_bad['s_BA'][elec]
            s_BA = np.vstack((s_BA, BA)) if np.size(s_BA) else BA
            label = pow_mat_bad['labels'][elec]
            #label_num = label+'_'+str(elec)
            channel = pow_mat_bad['channels'][elec]               
            s_labels = np.vstack((s_labels, label)) if np.size(s_labels) else label
            s_channels = np.vstack((s_channels, channel)) if np.size(s_channels) else channel
            ### POWER values for all electrodes for 1 freq for Good & Bad conditions
            elec_pow_bad = pow_mat_bad['xpow'][j,elec,:-6,:]
            elec_pow_good = pow_mat_good['xpow'][j,elec,:-6,:]
            #print(elec_pow_bad.shape, elec_pow_good.shape)
            ## AVERAGE POWER power across trials + select values only AFTER odor perception (14 windows)
            elec_pow_bad = np.mean(elec_pow_bad, axis=1)
            elec_pow_good = np.mean(elec_pow_good, axis=1)
            s_elec_pow_bad = np.vstack((s_elec_pow_bad, elec_pow_bad)) if np.size(s_elec_pow_bad) else elec_pow_bad
            s_elec_pow_good = np.vstack((s_elec_pow_good, elec_pow_good)) if np.size(s_elec_pow_good) else elec_pow_good
            s_elec = np.vstack((s_elec,str(elec))) if np.size(s_elec) else str(elec)
            ### STATS & DECODING ACCURACY for all electrodes
            da = np.mean(np.load(join(path_classif, freq+'/auc/'+su+'_auc_Good_Bad_'+label+'_('+str(elec)+').npy')),axis=0)
            th_05 = np.load(join(path_classif, freq+'/auc/'+su+'_th_0_05_perm_'+label+'_('+str(elec)+').npy'))
            th_01 = np.load(join(path_classif, freq+'/auc/'+su+'_th_0_01_perm_'+label+'_('+str(elec)+').npy'))
            #print('for one elec', stats.shape, da.shape,th_05.shape, th_01.shape)
            ### AVERAGE stats across repetitions for da + select only values AFTER odor perception
            s_da = np.vstack((s_da, da)) if np.size(s_da) else da
            s_th_05 = np.vstack((s_th_05, th_05)) if np.size(s_th_05) else th_05
            s_th_01 = np.vstack((s_th_01, th_01)) if np.size(s_th_01) else th_01
        #print(s_xyz.shape)
    print ('before', 'pow good',s_elec_pow_good.shape, 'pow bad',s_elec_pow_bad.shape,
           'da', s_da.shape, 'th',s_th_05.shape, s_th_01.shape,'aal',s_AAL.shape,'BA',s_BA.shape)   
    su_codes = np.squeeze(su_codes)
    s_labels = np.squeeze(s_labels)
    s_AAL = np.squeeze(s_AAL)
    s_AAL_RL = np.squeeze(s_AAL_RL)
    s_BA = np.squeeze(s_BA)
    s_channels = np.squeeze(s_channels)
    s_elec = np.squeeze(s_elec)
    s_da = np.squeeze(s_da)
    s_th_05 = np.squeeze(s_th_05)
    s_th_01 = np.squeeze(s_th_01)
    print ('after', 'pow good',s_elec_pow_good.shape, 'pow bad',s_elec_pow_bad.shape,
           'da', s_da.shape, 'th',s_th_05.shape, s_th_01.shape,'aal',s_AAL.shape, 'BA',s_BA.shape)
    su_coord_label = {'su_codes':su_codes, 's_xyz':s_xyz, 's_labels':s_labels,'s_channels':s_channels, 
                      's_elec':s_elec, 's_elec_pow_bad':s_elec_pow_bad,
                     's_elec_pow_good':s_elec_pow_good,'s_da':s_da,'s_th_05':s_th_05, 
                      's_th_01':s_th_01,'s_aal':s_AAL, 's_BA':s_BA, 's_aal_RL':s_AAL_RL}
    file_source = 'All_subjects_sources_'+freq+'_'+phase+'_RecScore_Expi_AAL_BA_'+str(bsl)+'.npz'
    np.savez(path2save+file_source,**su_coord_label)
```

```python
from brainpipe.system import study
import numpy as np
from os import listdir
from os.path import isfile, join
import numpy as np

## Find strings between two substrings
def find_between( s, first, last ):
    try:
        start = s.index(first) + len(first)
        end = s.index(last, start)
        return s[start:end]
    except ValueError:
        return ""

## CREATE A DICTIONARY OF ALL ELECTRODES SIGNIFICANT in a specific folder
st = study('Olfacto')
bsl = 'None'
phases = ['rest','odor']

for phase in phases:
    path_classif = join(st.path, 'classified/1_Classif_Power_EpiScore_all_electrodes_win700_step100_Rest_Percept/All_balanced_1_100perm_DA_stats_sametrials_'+str(bsl)+'/'+phase+'/')
    path_pow = join(st.path, 'feature/7_Power_E1E2_Odor_Good_Bad_EpiScore_Rest_Percept/')
    path2save = join(st.path, 'Visbrain/npz_files_rest_odor/')
    subjects = ['CHAF','FERJ','VACJ','SEMC','PIRJ','LEFC','MICP'] #respectively S0 S1 S2 S3 S4 S5
    freqs = ['0_VLFC', '1_delta', '2_theta', '3_alpha', '4_beta','5_gamma1','6_gamma2','7_gamma3']

    for j,freq in enumerate(freqs):
        path_data = join(path_classif, freq+'/fig')
        # Get dict of electrodes list by subject
        all_files = [f for f in listdir(path_data) if isfile(join(path_data, f))]
        su_elecs = {}
        for su in subjects:
            listelec_su = []
            d = {su : listelec_su}
            su_elecs.update(d)
            for file in all_files:
                if file.startswith(su):
                    elec_num = int(find_between(file, "(", ")" ))
                    listelec_su.append(elec_num)

    ## Create a npz file with all sources informations + POWER (subjects, coords, labels)
        su_codes, s_xyz, s_labels, s_elec_num_su, s_elec_pow_bad, s_elec_pow_good = np.array([]), np.array([]), np.array([]),np.array([]),np.array([]),np.array([])
        s_stats, s_stats_t0, s_da, s_th_05, s_th_01 = np.array([]),np.array([]),np.array([]),np.array([]),np.array([])
        for i,su in enumerate(subjects):
            for elec in su_elecs[su]:
                ### INFO electrodes for all subjects
                su_codes = np.vstack((su_codes,'S'+str(i))) if np.size(su_codes) else 'S'+str(i)
                xyz = np.load(path_pow+su+'_concat_odor_bad_bipo_new_'+str(bsl)+'_power.npz')['xyz'][elec]
                s_xyz = np.vstack((s_xyz, xyz)) if np.size(s_xyz) else xyz
                label = np.load(path_pow+su+'_concat_odor_bad_bipo_new_'+str(bsl)+'_power.npz')['labels'][elec]
                label_num = label+'_'+str(elec)
                s_labels = np.vstack((s_labels, label_num)) if np.size(s_labels) else label_num
                ### POWER values for all electrodes for 1 freq for Good & Bad conditions
                elec_pow_bad = np.load(path_pow+su+'_concat_'+phase+'_bad_bipo_new_'+str(bsl)+'_power.npz')['xpow'][j][elec]
                elec_pow_good = np.load(path_pow+su+'_concat_'+phase+'_good_bipo_new_'+str(bsl)+'_power.npz')['xpow'][j][elec]
                ## AVERAGE POWER power across trials + select values only AFTER odor perception (14 windows)
                elec_pow_bad = np.mean(elec_pow_bad, axis=1)
                elec_pow_good = np.mean(elec_pow_good, axis=1)
                s_elec_pow_bad = np.vstack((s_elec_pow_bad, elec_pow_bad)) if np.size(s_elec_pow_bad) else elec_pow_bad
                s_elec_pow_good = np.vstack((s_elec_pow_good, elec_pow_good)) if np.size(s_elec_pow_good) else elec_pow_good
                s_elec_num_su = np.vstack((s_elec_num_su,su+'_'+str(elec))) if np.size(s_elec_num_su) else su+'_'+str(elec)
                ### STATS & DECODING ACCURACY for all electrodes
                stats = np.load(join(path_classif, freq+'/stat/'+su+'_pvals_Epi_'+label+'_('+str(elec)+').npy'))
                stats_t0 = np.load(join(path_classif, freq+'/stat/'+su+'_t0_Epi_'+label+'_('+str(elec)+').npy'))
                da = np.mean(np.load(join(path_classif, freq+'/da/'+su+'_da_Epi_'+label+'_('+str(elec)+').npy')),axis=0)
                th_05 = np.load(join(path_classif, freq+'/da/'+su+'_th_0_05_perm_Epi_'+label+'_('+str(elec)+').npy'))
                th_01 = np.load(join(path_classif, freq+'/da/'+su+'_th_0_01_perm_Epi_'+label+'_('+str(elec)+').npy'))
                #print('for one elec', stats.shape, da.shape,th_05.shape, th_01.shape)
                ### AVERAGE stats across repetitions for da + select only values AFTER odor perception
                s_stats = np.vstack((s_stats, stats)) if np.size(s_stats) else stats
                s_stats_t0 = np.vstack((s_stats_t0, stats_t0)) if np.size(s_stats_t0) else stats_t0
                s_da = np.vstack((s_da, da)) if np.size(s_da) else da
                s_th_05 = np.vstack((s_th_05, th_05)) if np.size(s_th_05) else th_05
                s_th_01 = np.vstack((s_th_01, th_01)) if np.size(s_th_01) else th_01
        print ('before', 'pow good',s_elec_pow_good.shape, 'pow bad',s_elec_pow_bad.shape,
               'stats', s_stats.shape, 't0', s_stats_t0.shape, 'da', s_da.shape, 'th',s_th_05.shape, s_th_01.shape)   
        su_codes = np.squeeze(su_codes)
        s_labels = np.squeeze(s_labels)
        s_elec_num_su = np.squeeze(s_elec_num_su)
        s_stats = np.squeeze(s_stats)
        s_stats_t0 = np.squeeze(s_stats_t0)
        s_da = np.squeeze(s_da)
        s_th_05 = np.squeeze(s_th_05)
        s_th_01 = np.squeeze(s_th_01)
        print ('after', 'pow good',s_elec_pow_good.shape, 'pow bad',s_elec_pow_bad.shape,
               'stats', s_stats.shape, 'da', s_da.shape, 'th',s_th_05.shape, s_th_01.shape)
        su_coord_label = {'su_codes':su_codes, 's_xyz':s_xyz, 's_labels':s_labels, 
                          's_elec_num_su':s_elec_num_su, 's_elec_pow_bad':s_elec_pow_bad,
                         's_elec_pow_good':s_elec_pow_good,'s_stats':s_stats,'s_stats_t0':s_stats_t0,
                          's_da':s_da,'s_th_05':s_th_05, 's_th_01':s_th_01}
        file_source = 'All_sources_'+freq+'_'+phase+'_EpiScore_classif_'+str(bsl)+'.npz'
        np.savez(path2save+file_source,**su_coord_label)
```
