```python
from brainpipe.system import study
import numpy as np
from os import listdir
from os.path import isfile, join
import numpy as np

st = study('Olfacto')
phase = 'odor'
path_classif = join(st.path, 'classified/1_Classif_Phase_EpiScore_sel_electrodes_win700_step100_expi/')
path_phase = join(st.path, 'feature/7_Phase_E1E2_Odor_Good_Bad_EpiScore_Expi/')
path2save = join(st.path, 'Visbrain/npz_files_odor_expi/Phase/')
subjects = ['LEFC','CHAF','VACJ','SEMC','FERJ','MICP','PIRJ'] #respectively S0 S1 S2 S3 S4 S5
freqs = ['0_VLFC', '1_delta', '2_theta', '3_alpha']

## Create a npz file with all sources informations + PHASE (subjects, coords, labels)
for j,freq in enumerate(freqs):
    su_codes, s_xyz, s_channels,s_labels, s_elec, s_elec_phase_bad, s_elec_phase_good = np.array([]),np.array([]), np.array([]), np.array([]),np.array([]),np.array([]),np.array([])
    s_stats, s_stats_t0, s_da, s_th_05, s_th_01 = np.array([]),np.array([]),np.array([]),np.array([]),np.array([])
    print(freq)
    for i,su in enumerate(subjects):
        print(su)
        xyz_all = np.load(path_phase+su+'_odor_bad_bipo_sel_phase.npz')['xyz']
        for elec in range(xyz_all.shape[0]):
            ### INFO electrodes for all subjects
            su_codes = np.vstack((su_codes,'S'+str(i))) if np.size(su_codes) else 'S'+str(i)
            xyz = xyz_all[elec]
            s_xyz = np.vstack((s_xyz, xyz)) if np.size(s_xyz) else xyz
            label = np.load(path_phase+su+'_odor_bad_bipo_sel_phase.npz')['labels'][elec]
            label_num = label+'_'+str(elec)
            channel = np.load(path_phase+su+'_odor_bad_bipo_sel_phase.npz')['channels'][elec]               
            s_labels = np.vstack((s_labels, label_num)) if np.size(s_labels) else label_num
            s_channels = np.vstack((s_channels, channel)) if np.size(s_channels) else channel
            ### PHASE values for all electrodes for 1 freq for Good & Bad conditions
            elec_phase_bad = np.load(path_phase+su+'_odor_bad_bipo_sel_phase.npz')['phase'][elec][j]
            elec_phase_good = np.load(path_phase+su+'_odor_good_bipo_sel_phase.npz')['phase'][elec][j]
            ## AVERAGE PHASE across trials
            elec_phase_bad = np.mean(elec_phase_bad, axis=1)
            elec_phase_good = np.mean(elec_phase_good, axis=1)
            s_elec_phase_bad = np.vstack((s_elec_phase_bad, elec_phase_bad)) if np.size(s_elec_phase_bad) else elec_phase_bad
            s_elec_phase_good = np.vstack((s_elec_phase_good, elec_phase_good)) if np.size(s_elec_phase_good) else elec_phase_good
            s_elec = np.vstack((s_elec,str(elec))) if np.size(s_elec) else str(elec)
            ### STATS & DECODING ACCURACY for all electrodes
            da = np.mean(np.load(join(path_classif, su+'/'+freq+'/auc/'+su+freq[1:]+'_phase_auc_Good_Bad_elec_'+channel+'_'+label+'_('+str(elec)+').npy')),axis=0)
            th_05 = np.load(join(path_classif, su+'/'+freq+'/auc/'+su+freq[1:]+'_phase_th_0_05_perm_Good_Bad_elec_'+channel+'_'+label+'_('+str(elec)+').npy'))
            th_01 = np.load(join(path_classif, su+'/'+freq+'/auc/'+su+freq[1:]+'_phase_th_0_01_perm_Good_Bad_elec_'+channel+'_'+label+'_('+str(elec)+').npy'))
            #print('for one elec', stats.shape, da.shape,th_05.shape, th_01.shape)
            ### AVERAGE stats across repetitions for da + select only values AFTER odor perception
            s_da = np.vstack((s_da, da)) if np.size(s_da) else da
            s_th_05 = np.vstack((s_th_05, th_05)) if np.size(s_th_05) else th_05
            s_th_01 = np.vstack((s_th_01, th_01)) if np.size(s_th_01) else th_01
        print(s_xyz.shape)
    print ('before', 'pow good',s_elec_phase_good.shape, 'pow bad',s_elec_phase_bad.shape,
           'da', s_da.shape, 'th',s_th_05.shape, s_th_01.shape)   
    su_codes = np.squeeze(su_codes)
    s_labels = np.squeeze(s_labels)
    s_channels = np.squeeze(s_channels)
    s_elec = np.squeeze(s_elec)
    s_da = np.squeeze(s_da)
    s_th_05 = np.squeeze(s_th_05)
    s_th_01 = np.squeeze(s_th_01)
    print ('after', 'pow good',s_elec_phase_good.shape, 'pow bad',s_elec_phase_bad.shape,
           'da', s_da.shape, 'th',s_th_05.shape, s_th_01.shape)
    su_coord_label = {'su_codes':su_codes, 's_xyz':s_xyz, 's_labels':s_labels,'s_channels':s_channels, 
                      's_elec':s_elec, 's_elec_phase_bad':s_elec_phase_bad,
                     's_elec_phase_good':s_elec_phase_good,'s_da':s_da,'s_th_05':s_th_05, 
                      's_th_01':s_th_01}
    file_source = 'All_subjects_sources_'+freq+'_'+phase+'_EpiScore_Expi.npz'
    np.savez(path2save+file_source,**su_coord_label)
```
