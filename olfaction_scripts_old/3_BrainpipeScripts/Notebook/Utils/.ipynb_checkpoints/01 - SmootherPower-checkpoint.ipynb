### Compute the power :

```python
fc = (8,13)
baseline, norm = (250,750), 0
width, step = 700, 50#700, 50
split = None
fcname = str(fc[0])+'-'+str(fc[1])

powObj = power(sf, 4001, f=fc, baseline=baseline, width=width, step=step, split=split,
               method='hilbert1', norm=norm, time=n.arange(4001)-1000)
# time = n.array(powObj.xvec).astype(int)-1000


n.array(powObj._window[5]).mean()
xPower = powObj.get(x, n_jobs=-1)[0]
# time = powObj.xvec
```

```python
x_m =  n.tile(n.mean(xPower.T, 0), (xPower.T.shape[0], 1))
x_m.shape
xPower = n.squeeze(xPower)
```

### Classify each time point :

```python
clfObj = classify(y, clf='svm', cvtype='sss', cvArg={'rep':10}, clfArg={'kern':'rbf'})
# da = clfObj.fit(xPower, n_jobs=-1)
da = clfObj.fit(xPower, n_jobs=-1, center=True)[0]#, n_perm=20, method='label_rnd') , pvalue, daperm
```

```python
da.max(), da.mean(1).argmax(), powObj._window[da.mean(1).argmax()], da.mean(0).max()
```

### Border Plot :

```python
from matplotlib.ticker import ScalarFormatter, MaxNLocator
# xPower[:,83], xPower[:,93] = xPower[:,82], xPower[:,92]
# xPower = n.delete(xPower,(83,91,92,93,94),axis=1)
# y = n.delete(n.array(y),(83,91,92,93,94),axis=0)

titre=sujet#+'-'+main+', elec:'+channel[0][elec][0]+'('+str(elec)+'), f='+str(fc)+', clf='+clfObj.clf.shStr

xfmt = ScalarFormatter(useMathText=True)
xfmt.set_powerlimits((0,3))

fig = plt.figure(1,figsize=(10,8))
titre='Gamma power directions decoding (BA6 - S6)'

# Plot the phase features :
plt.subplot(211)
BorderPlot(powObj.xvec,xPower,y=y, xlabel='',title=titre,legend=['Up','Right','Down','Left'],
           alpha=0.3,linewidth=4,color=['b','g','r','darkorange'],xticks=[0,1500],ylabel=r'$\mu$VÂ² / hz',
           kind='sem')
rmaxis(plt.gca(), ['right', 'top'])
addLines(plt.gca(),vLines=[0,1500],vWidth=[1,1],vShape=['-','-'])
# plt.gca().yaxis.set_major_locator(MaxNLocator(3,integer=True))
plt.gca().yaxis.set_major_formatter(xfmt)

plt.legend(loc=0, handletextpad=0.1, frameon=False)

plt.subplot(212)
BorderPlot(powObj.xvec,da,color='darkslateblue',kind='std',ylim=[20,50],xlabel='Time (ms)',
           ylabel='Decoding accuracy (%)',xticks=[0,1500],linewidth=4,alpha=0.3)
rmaxis(plt.gca(), ['right', 'top'])
addLines(plt.gca(),vLines=[0,1500],vWidth=[1,1],hLines=[25],vShape=['-','-'],hWidth=[2])
plt.plot(powObj.xvec, 39.8*n.ones((len(powObj.xvec),1)), '--', color='dimgrey', linewidth=2, label='p<.05')
# plt.plot(time, daperm.max()*n.ones((len(time),1)), '--', color='dimgrey', linewidth=2, label='p<.05')
plt.legend(loc=2, handletextpad=0.1, frameon=False)

fig.savefig('C_rev-RH_elec73_power_alpha.png', dpi=300, bbox_inches='tight')
```
