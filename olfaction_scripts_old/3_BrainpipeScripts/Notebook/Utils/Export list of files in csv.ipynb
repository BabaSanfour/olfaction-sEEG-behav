```python
import numpy as np
import pandas as pd
import scipy.io as sio
from os import path
from os import listdir
```

```python
path_all = '/media/karim/Seagate Backup Plus Drive/Memory_Laughter_Videos/'

folds = ['Neutral_Edited', 'Laughter_Edited', 'Funny_Edited']

for fold in folds:
    path_names = path_all+fold
    print(path_names)
    all_files = [f for f in listdir(path_names) if path.isfile(path.join(path_names, f))]
    print(all_files)
    d = {'Videos_'+fold: all_files}
    df = pd.DataFrame(data=d)
    print(df)
    df.to_csv(path_all+'all_videos_'+fold+'.csv', sep=',')
```
