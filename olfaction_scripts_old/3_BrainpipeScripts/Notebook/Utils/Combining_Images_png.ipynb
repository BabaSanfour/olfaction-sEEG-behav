## Combining pictures into one

### Avec matplotlib

```python
import numpy as np
from os import path, listdir
from brainpipe.system import study
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#s%matplotlib inline

st = study('Olfacto')
path_images = path.join(st.path, 'TimeFrequency/TF_Encoding_all_filter1_th40_400art_30_250_GoodBad_RECSCORE/Stat_uncorrected_learning_OK/')
path2save = path.join(path_images, 'TF_Learning_Grouped/')

subjects = ['LEFC','MICP','VACJ','PIRJ','CHAF','SEMC']

# n_elec = {
# 'CHAF' : 107,
# 'VACJ' : 139, 
# 'SEMC' : 107,
# 'PIRJ' : 106,
# 'LEFC' : 193,
# 'MICP' : 105,
#             }

n_elec = {
'CHAF' : [36,95],
'VACJ' : [3,61], 
'SEMC' : [0,6,25],
'PIRJ' : [96],
'LEFC' : [0,6,88,91],
'MICP' : [88],
            }

for su in subjects:
    for elec in n_elec[su]:
        plt.figure(figsize=(20,10))
        files = st.search(su+'_TF_Odor_Elec_'+str(elec)+'_', folder=path_images)
        for file in files:
            if file.endswith("Learning_0.png"): 
                img_1 = mpimg.imread(path_images + file)
            elif file.endswith("Learning_1.png"): 
                img_2 = mpimg.imread(path_images + file)
            elif file.endswith("Learning_2.png"): 
                img_3 = mpimg.imread(path_images + file)

        plt.subplot(1, 3, 1)
        plt.imshow(img_1)
        plt.axis('off')
        plt.subplot(1, 3, 2)
        plt.imshow(img_2)
        plt.axis('off')
        plt.subplot(1, 3, 3)
        plt.imshow(img_3)
        plt.axis('off')

        savename = (path2save+ files[0].replace('Learning_1.png', 'Learning_all.png'))
        print ('saving --> ',savename)
        plt.savefig(savename, dpi=300, bbox_inches='tight')
        plt.clf()
        plt.close('all')
    del elec, img_1, img_2, img_3, files
```

```python
import numpy as np
from os import path, listdir
from brainpipe.system import study
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#s%matplotlib inline

st = study('Olfacto')
path_labels = path.join(st.path, 'database/TS_E_all_by_odor_th40_art400_30_250_Good_Bad_RecScore')
path_images = path.join(st.path, 'feature/8_PLV_RECSCORE_Good_Bad_across_time_subset/')
path2save = path.join(path_images, 'PLV_grouped/')

subjects = ['VACJ','LEFC','MICP','PIRJ','SEMC'] #'CHAF'

n_elec = {
'VACJ' : [1,2,3,11,12,13,14,15,16,17,22,23,24,60,61,62],
'SEMC' : [0,1,2,3,4],
'PIRJ' : [0,1,2,3,4,11,12,13,14,15,16,22,23,24,25,26,33,34,35,36,37,38],
'LEFC' : [0,1,11,12,13,14,22,23,24,25,26,27],
'MICP' : [0,1,2,3,9,18,10,11,12,19,20,21,22,23,29,30,31,32,33,40,41,42,43,44],
    }

freqs = ['delta', 'theta', 'alpha', 'beta', 'gamma30-60', 'gamma60-150']
conds = ['bad', 'good']

for su in subjects:
    for elec in n_elec[su]:
        label = np.load(path.join(path_labels, su+'_concat_odor_bad_bipo.npz'))['label'][elec]
        print(label)
        for freq in freqs:
            file1 = st.search(su+'_'+conds[0]+'_plv_'+freq+'_elec_('+str(elec)+')_'+label+'.png', folder=path_images)[0]
            print(file1)
            img_1 = mpimg.imread(path_images + file1)
            file2 = st.search(su+'_'+conds[1]+'_plv_'+freq+'_elec_('+str(elec)+')_'+label+'.png', folder=path_images)[0]
            print(file2)
            img_2 = mpimg.imread(path_images + file2)
            plt.figure(figsize=(20,10))
            plt.subplot(1, 2, 1)
            plt.imshow(img_1)
            plt.axis('off')
            plt.subplot(1, 2, 2)
            plt.imshow(img_2)
            plt.axis('off')

            savename = (path2save+ file1.replace('good_plv', 'good_bad_plv'))
            print ('saving --> ',savename)
            plt.savefig(savename, dpi=300, bbox_inches='tight')
            plt.clf()
            plt.close('all')
```

```python
import numpy as np
from os import path, listdir
from brainpipe.system import study
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#s%matplotlib inline

st = study('Olfacto')
path_images = path.join(st.path, 'Visbrain/Pow_Signif_ROI/')
path2save = path.join(path_images,'Groupes/' )
print(path2save)

freqnames = ['0_theta']
conds = ['bad','good']
wins = ['win_0', 'win_1', 'win_2', 'win_3', 'win_4', 'win_5']
rotations = ['coronal_0','coronal_1']

for freq in freqnames:
    for win in wins:
        plt.figure(figsize=(12,10))
        files = st.search('Power_'+freq+'_'+win+'_', folder=path_images)
        for file in files:
            if file.endswith('coronal_0_bad.bmp'): 
                img_1 = mpimg.imread(path_images + file)
            elif file.endswith('coronal_0_good.bmp'): 
                img_2 = mpimg.imread(path_images + file)
            elif file.endswith('coronal_1_bad.bmp'): 
                img_3 = mpimg.imread(path_images + file)
            elif file.endswith('coronal_1_good.bmp'): 
                img_4 = mpimg.imread(path_images + file)
                
        plt.subplot(2, 2, 1)
        plt.imshow(img_1)
        plt.title('Power '+freq+' '+win+' bad Front View')
        plt.axis('off')
        plt.subplot(2, 2, 2)
        plt.title('Power '+freq+' '+win+' good Front View')
        plt.imshow(img_2)
        plt.axis('off')
        plt.subplot(2, 2, 3)
        plt.title('Power '+freq+' '+win+' bad Back View')
        plt.imshow(img_3)
        plt.axis('off')
        plt.subplot(2, 2, 4)
        plt.title('Power '+freq+' '+win+' good Back View')
        plt.imshow(img_4)
        plt.axis('off')

        savename = (path2save+ files[0].replace('coronal_0_bad.bmp', 'coronal_bad_good.jpg'))
        print ('saving --> ',savename)
        plt.savefig(savename, dpi=300, bbox_inches='tight')
        plt.clf()
        plt.close('all')
    del img_1, img_2,files
```

### combine right and left DA 

```python
import numpy as np
from os import path, listdir
from brainpipe.system import study
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#s%matplotlib inline

st = study('Olfacto')
path_images = path.join(st.path, 'Visbrain/')
path2save = path.join(path_images, 'Final/')

freqnames = ['theta', 'alpha', 'beta', 'gamma30-60', 'gamma60-150']

for freq in range(0,5):
    for win in range(7):
        plt.figure(figsize=(12,10))
        files = st.search('DA_'+freqnames[freq]+'_all_subjects_window'+str(win)+'_', folder=path_images)
        print ('DA_'+freqnames[freq]+'_all_subjects_window'+str(win), files)
        for file in files:
            if file.endswith("sagittal_0.jpg"): 
                img_1 = mpimg.imread(path_images + file)
            elif file.endswith("sagittal_1.jpg"): 
                img_2 = mpimg.imread(path_images + file)

        plt.subplot(1, 2, 1)
        plt.imshow(img_1)
        plt.title('DA_'+freqnames[freq]+'_window'+str(win)+'_left')
        plt.axis('off')
        plt.subplot(1, 2, 2)
        plt.title('DA_'+freqnames[freq]+'_window'+str(win)+'_right')
        plt.imshow(img_2)
        plt.axis('off')

        savename = (path2save+ files[0].replace('sagittal_0', 'sagittal'))
        print ('saving --> ',savename)
        plt.savefig(savename, dpi=300, bbox_inches='tight')
        plt.clf()
        plt.close('all')
    del img_1, img_2,files
```
