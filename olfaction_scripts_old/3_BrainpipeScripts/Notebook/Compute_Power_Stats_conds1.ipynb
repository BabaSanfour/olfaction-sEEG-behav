# Compute Power and plot for conditions

## Compute stats and power for one electrode

```python
import numpy as np
import matplotlib.pyplot as plt
from os import path

#%matplotlib notebook

from brainpipe.system import study
from brainpipe.feature import power, TF
from brainpipe.visual import *
from brainpipe.statistics import *

if __name__ == '__main__':
    # Define which condition / electrode
    st = study('Olfacto')
    subjects = ['CHAF', 'SEMC', 'FERJ', 'VACJ',]
    #trig = ['old', 'new']
    trig = ['hit', 'cr']

    n_elec = {
            'CHAF' : 108,
            'SEMC' : 107,
            'VACJ' : 137,
            'FERJ' : 108,
        }

#     freq_bands = {
#             'delta' : [2, 4],
#             'theta' : [5, 7],
#             'alpha' : [8, 13],
#             'beta' : [13, 30],
#             'gamma30-60' : [30, 60],
#             'gamma60-120' :[60, 120],
#                     }  

    # Get all files include in the database
    st = study('Olfacto')
    #files = st.search('_bipo.npz', folder='database/TS_R_Rec_cond_by_block_trigs_filter1_500art')

    for su in subjects:
        for elec in range(0, n_elec[su], 1):
            for tr in trig:
                file = su+'_'+tr+'_R1R2R3_concat_allfilter1_bipo.npz'
                mat = np.load(path.join(st.path, 'database/TS_R_Rec_cond_by_block_trigs_filter1_500art', file))
                x, channel, sf = mat['x'], list(mat['channel']), mat['sf']
                statmeth = 'permutation'
                f = [8, 13]
                fname = 'alpha'
                print (su, elec, tr)

                # Calcul de la puissance :
                powObj = power(sf, x.shape[1], f=f, norm=3, baseline=(0, 512), width=100, step=50)
                print (x.shape[1])
                xpow, pval = powObj.get(x[elec, ...], statmeth=statmeth, n_perm = 1000, maxstat=True, n_jobs=1)
                print (xpow.shape)

                # Juste pour enlever les dimensions inutiles :
                xpow, pval = np.squeeze(xpow), np.squeeze(pval)
                time = powObj.xvec

                # Plot the TF maps :
                plt.figure(0)

                if tr == 'hit':
                    BorderPlot(time, xpow, color='darkblue')
                    addPval(plt.gca(), pval, p=0.05, x=time, y=0, color='darkred', lw=3)
                    addPval(plt.gca(), pval, p=0.01, x=time, y=0.5, color='darkgreen', lw=4)

                elif tr == 'cr':
                    BorderPlot(time, xpow, color='darkred')
                    addPval(plt.gca(), pval, p=0.05, x=time, y=0, color='darkred', lw=3)
                    addPval(plt.gca(), pval, p=0.01, x=time, y=0.5, color='darkgreen', lw=4)

                #Save all your plots
                rep = path.join(st.path, 'feature/Stat_Power_Hit_Cr_alpha',su)
                print (rep)
                fname = (rep + '_'+tr+'_R1R2R3_concat_bipo_' + channel [elec] +'_'+fname+'.png')
                print (fname)
                plt.savefig(fname, dpi=300, bbox_inches='tight')
                plt.close()
            del x, sf, channel
```
