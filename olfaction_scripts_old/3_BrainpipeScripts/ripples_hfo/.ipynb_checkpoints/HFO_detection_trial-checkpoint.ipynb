```python
import os, requests, tempfile, pickle, time
from os import path, listdir, makedirs
from pyhfo_detect.io import add_metadata
from pyhfo_detect.core import (ll_detect, rms_detect, morphology_detect,
                               cs_detect_beta, hilbert_detector_1_0)
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib notebook

from brainpipe.system import study
```

```python
st = study('Olfacto')
data_path = path.join(st.path,'database/Encoding_EpiPerf_LowHigh/')
filename = data_path+'{}_odor_{}_bipo_sel_physFT.npz'

su = 'FERJ'
conds = ['low','high']
low_mat = np.load(filename.format(su,conds[0]),allow_pickle=True)
high_mat = np.load(filename.format(su,conds[1]))
print(low_mat.files)
```

```python
data_low = low_mat['x'][:,1024:2304] #from -500ms to +2000ms
data_high = high_mat['x'][:,1024:2304] #from -500ms to +2000ms
nelecs, npts, ntrials_low = data_low.shape
ntrials_high = data_high.shape[2]
channels = low_mat['channels']
labels = low_mat['Mai_RL']
sf = low_mat['sf']
dur_min = round(25*sf/1000)
isi_min = round(15*sf/1000)

# df_global = pd.DataFrame(columns=['subject','elec','trial','duration','ripples'])
for elec in range(nelecs):
    low_freq = []
    for trial in range(ntrials_low):
#         r = (elec*ntrials)+trial
        data_elec = data_low[elec,:,trial]
        # detect and save HFO for one electrode
        RMS_df = rms_detect(data_elec, sf, 80, 120, 3, 0.1, 0.25)
        
        #remove too short ripples from the detection list
        RMS_df['delta_time'] = RMS_df['event_stop'] - RMS_df['event_start']
        RMS_df = RMS_df.loc[RMS_df.duration > dur_min & RMS_df.delta_time > isi_min]
        low_freq.append(RMS_df.shape[0])
    
    high_freq = []
    for trial in range(ntrials_high):
#         r = (elec*ntrials)+trial
        data_elec = data_high[elec,:,trial]
        # detect and save HFO for one electrode
        RMS_df = rms_detect(data_elec, sf, 80, 120, 3, 0.1, 0.25)
        
        #remove too short ripples from the detection list
        RMS_df['duration'] = RMS_df['event_stop'] - RMS_df['event_start']
        RMS_df['delta_ripple'] = RMS_df['event_start'].shift(-1) - RMS_df['event_start']
        RMS_df = RMS_df.loc[RMS_df.duration > dur_min & RMS_df.delta_time > isi_min]
        high_freq.append(RMS_df.shape[0])
    
    #test whether the number of ripples is significantly different
    
```
