## 1-Concatenate ODOR & REST
    A lancer pour générer le fichier "group_files" nécessaire pour grouper les 
    odeurs en différentes catégories

```python
import numpy as np
import sys, os

path_data = r'/media/karim/Datas4To/'

reps = [#'TS_E_all_cond_by_block_trigs_odors/',
        'TS_R_all_cond_by_block_trigs_odors/',
        'TS_R_all_cond_by_block_trigs_rec/']
subjects = ['PIRJ','CHAF','VACJ','SEMC','LEFC','FERJ']
runs = ['R1', 'R2', 'R3']
#runs = ['E1', 'E2']
#conds = ['all']
#conds = ['hit', 'cr', 'fa', 'miss']
conds = range(1,21) #['no_odor']

for rep in reps:
    ##create the repository if necessary
    path2save = os.path.join(path_data+rep,'_group_files')
    try:
        os.makedirs(path2save)
    except OSError:
        if os.path.exists(path2save):
            pass
        else:
            raise

    ##Concatenate Odors and Rest for each Run
    for su in subjects:
        for cond in conds:
            concat_runs = np.array([])
            for run in runs:
                path = os.path.join(path_data+rep, su, run)
                filename = os.path.join(path,'correct_ts_'+str(cond)+'_by_odor_trigs.npy' )
                mat = np.load(filename)
                print (su, run, 'odor #',cond,mat.shape)
                if mat.size == 0: 
                    print ('fichier vide')
                else :
                    concat_runs = np.vstack((concat_runs,mat)) if np.size(concat_runs) else mat
            filename_odor = (su+'_odor_'+str(cond)+'.npy')
            np.save((os.path.join(path2save,filename_odor)), concat_runs)
            print(su,'concat shape', concat_runs.shape)
```

## 2- Create .mat matrix for learning E1 E2 BY ODOR

```python
import numpy as np
from os import path, makedirs
import pandas as pd
from scipy.io import savemat

path_data = r'/media/karim/Datas4To/'
reps = [#'TS_R_all_cond_by_block_trigs_odors/',
        'TS_R_all_cond_by_block_trigs_rec/']
#reps = ['TS_E_all_cond_by_block_trigs_odors/']
subjects = ['PIRJ','CHAF','VACJ','SEMC','LEFC','FERJ']
odors_all = list(range(1,21))
odors_su_E = {'CHAF' : ['1','2','4','5','3','8','7','9'],
                    'VACJ' : ['11','14','12','10','15','17','16','13'],
                    'SEMC' : ['7','10','11','12','13','5','8','9'],
                    'PIRJ' : ['1','9','5','4','6','7','18','15'],
                    'LEFC' : ['15','2','1','16','14','3','4','17'],
                    'FERJ' : ['7','12','16','17','13','1','5','2']}
odors_su_R = {   'CHAF': [6,18,19,20],
               'LEFC': [6,18,19,20],
               'PIRJ': [],
               'VACJ': [6,18,19,20],
               'SEMC': [6,18,19,20],
               'FERJ': [6,18,19,20],
               } #'MICP':(2,12,6,8,3,18,9,14)

for rep in reps:
    for su in subjects:
        odors_sel = list(set(odors_all)^set(odors_su_R[su]))
        for odor in odors_sel:
            print ('-> Computing on : ', su , 'cond', odor)
            dataname = path.join(path_data+rep,('_group_files/'+su+'_odor_'+str(odor)+'.npy'))
            data = np.load(dataname)
            print (su,'data shape', data.shape)
            sf = 512.

            ## Coordinates of electrodes included in Data
            elec_info = path.join(path_data+rep,(su+'correct_all_info.txt'))
            print(elec_info)
            loadfile = np.loadtxt(elec_info, dtype=str, delimiter=',')
            chanmonop = pd.DataFrame(loadfile,)
            coord = (np.array(chanmonop.iloc[:,0]), 
                     np.array(chanmonop.iloc[:,1]), 
                     np.array(chanmonop.iloc[:,2]), 
                     np.array(chanmonop.iloc[:,3]),
                     np.array(chanmonop.iloc[:,4]),)
            chanmonop_final = np.array(coord, dtype=object)

            ## Create and save the structure of np array (sig, sf, coord)
            path_save = path_data+rep+'_mat_by_odor/'
            if not path.exists(path_save):
                makedirs(path_save)
            savename = path.join(path_save, su+'_odor_'+str(odor)+'.mat')
            savemat(savename,{'x': data, 'chanmonop': chanmonop_final, 'sf': sf})
```
