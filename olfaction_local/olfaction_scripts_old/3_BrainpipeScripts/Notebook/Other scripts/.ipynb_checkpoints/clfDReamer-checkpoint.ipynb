```python
import numpy as np
from scipy.io import loadmat, savemat
from itertools import product
import matplotlib.pyplot as plt

%matplotlib notebook

from sklearn.model_selection import StratifiedKFold

from ipywksp import workspace
from brainpipe.classification import *
```

```python
#wk = workspace(theme="dark", autoHide=True)
```

```python
refDreamer = np.arange(18) + 1
refNnDreamer = np.array([19, 20] + [23, 24, 25, 26] + [k for k in range(28, 39)])
C = ['C1', 'C2']
path = "/home/karim/MATLAB/Work/Projet C1 C2/C1C2 j7 11/mean C1 C2/"
file = 'mean_{C}_s{s}.mat'


iteDream = product(C, refDreamer)
iteNnDream = product(C, refNnDreamer)

xDream, xNnDream = np.array([]), np.array([])
for C, name in iteDream:
    ck = loadmat(path+file.format(C=C, s=name))[C]

    if np.any(np.isnan(ck)):
        print(C, name)

    xDream = np.concatenate((xDream, ck), axis=0) if xDream.size else ck

for C, name in iteNnDream:
    ck = loadmat(path+file.format(C=C, s=name))[C]

    if np.any(np.isnan(ck)):
        print(C, name)

    xNnDream = np.concatenate((xNnDream, ck), axis=0) if xNnDream.size else ck


x = np.concatenate((xDream, xNnDream), axis=0)
y = np.array([0]*xDream.shape[0] + [1]*xNnDream.shape[0])
```

```python
x.shape
```

```python
print(x.shape, len(y))

# Define a pipeline :
cvout = StratifiedKFold(n_splits=10, shuffle=True)
mfp = MFpipe(y, cvout)

# name = 'svm_rbf_scale_kbest_fdr_optimized_C_gamma'
pipename = 'lda_scale_kbest'
mfp.default_pipeline(pipename, n_best=10)
da = mfp.fit(x, n_jobs=1, verbose=1)
```

```python
sf = mfp.selected_features()
sf["Ro"] = ["Test"]*10 + ["ok"]*9
sf["Ro2"] = ["Yjs"]*9 + ["iljjkij"]*10
```
